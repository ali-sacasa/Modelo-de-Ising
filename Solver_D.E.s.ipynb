## 1. Marco funcional y notación

Sea $N$ el número de espines. El espacio de Hilbert físico es

$$
H = (\mathbb{C}^2)^{\otimes N}, \quad \dim H = d := 2^N.
$$

Sea $H: H \to H$ el Hamiltoniano hermítico (por ejemplo de Ising).  El producto interno se denota $\langle \cdot, \cdot \rangle$ y la norma $\|\cdot\|$.

Objetivo: construir un subespacio de dimensión $M \ll d=2^N$, mediante una construcción geométrica basada en secciones holomorfas (estados coherentes), proyectar $H$ a ese subespacio y controlar el error espectral y dinámico, reduciendo las dimensiones y por ende, el procesamiento y memoria, siendo un sistema de ecuaciones linearizado más fácil de resolver.

---

## 2. Geometría compleja de $\mathbb{CP}^1$ y fibrado tautológico

### 2.1 Espacio proyectivo complejo

El espacio proyectivo complejo $\mathbb{C}P^n$ se define como el conjunto de líneas complejas que pasan por el origen en $\mathbb{C}^{n+1}$:

$$
\mathbb{C}P^n = \{ [z_0 : z_1 : \cdots : z_n] \,|\, (z_0,\ldots,z_n)\neq 0 \},
$$

donde $[z_0:...:z_n]$ denota la clase de equivalencia bajo la relación $(z_0,\ldots,z_n) \sim \lambda(z_0,\ldots,z_n)$ para $\lambda \in \mathbb{C}^\times$.

Cada punto de $\mathbb{C}P^n$ corresponde a una línea compleja unidimensional en $\mathbb{C}^{n+1}$.

### 2.2 Fibrado tautológico

El **fibrado tautológico** $\mathcal{O}(-1)$ sobre $\mathbb{C}P^n$ se define mediante

$$
\mathcal{O}(-1) = \{ ([z], v) \in \mathbb{C}P^n \times \mathbb{C}^{n+1} \, | \, v \in [z] \}.
$$

Su fibra en el punto $[z]$ es precisamente la línea compleja generada por $z$, de forma tal que, su dual dual $\mathcal{O}(1)$ es el **fibrado hiperbólico**, que tiene un rol central en la representación de estados cuánticos y secciones holomorfas.

---

## 3. Geometría Kähler y métrica de Fubini–Study

El espacio $\mathbb{C}P^n$ posee una estructura Kähler natural, con métrica inducida por el potencial de Kähler

$$
K(z, \bar{z}) = \log(1 + |z|^2),
$$

en coordenadas inhomogéneas $z = (z_1, \ldots, z_n)$. La forma de Kähler asociada, siendo la métrica de Fubini–Study es

$$
\omega_{FS} = i \partial \bar{\partial} K = \frac{i \sum_{i,j} [(1+|z|^2)\delta_{ij} - \bar{z}_i z_j]\, dz_i \wedge d\bar{z}_j}{(1+|z|^2)^2}.
$$

La integral de $\omega_{FS}$ sobre una 2-esfera compleja (por ejemplo $\mathbb{C}P^1$ embebida) da el **número de Chern** de $\mathcal{O}(1)$

$$
\frac{1}{2\pi}\int_{\mathbb{C}P^1}\omega_{FS} = 1.
$$

Esto establece el vínculo topológico entre los fibrados complejos y la cuantización geométrica. Utilizando de forma general una 2-forma diferencial $\omega=i g_{i\bar{j}}dz^i \wedge d\bar{z}^j$, el flujo de Ricci del sistema evolucionaria en función de un parámetro $\tau$ como


$$
\partial_{\tau} \omega=i\partial\bar{\partial} \ln{\det{g_{i\bar{j}}(\tau)}}.
$$

siendo de importancia para la renormalización geométrica y el estudio de las singularidades del espacio de estados.

## 4. Estados coherentes y representación holomorfa

Identificamos la esfera $S^2$ con $\mathbb{C}P^1$ mediante la proyección estereográfica vía un homeomorfismo diferenciable. Para cada $z \in \mathbb{C}$, el estado coherente (sección local de $\mathcal{O}(1)$) es

$$
|z\rangle = \frac{1}{\sqrt{1+|z|^2}}
\begin{pmatrix}
1 \\ z
\end{pmatrix} \in \mathbb{C}^2.
$$

Propiedades

$$
\langle z | z \rangle = 1, \quad 
\int_{\mathbb{C}} |z\rangle\langle z| \, \frac{2}{\pi}\frac{d^2z}{(1+|z|^2)^2} = I_2.
$$

Para $N$ espines

$$
|Z\rangle = \bigotimes_{i=1}^N |z_i\rangle, \quad
\int |Z\rangle \langle Z|\, d\mu^{(N)}(Z) = I_H.
$$

Por ejemplo, para un único espin, la ecuación se verifica expandiendo $|z\rangle\langle z|$ en la base $\{\ket{0},\ket{1}\}$ y usando coordenadas polares mediante la integral 

$$
\int_{0}^{2\pi} e^{im\phi}d\phi =0, \  \, \forall m \not= 0.
$$

La integral radial reproduce las constantes necesarias.

## 5. Kernel de Bergman y espacio de secciones holomorfas

Consideremos el espacio de secciones holomorfas de $\mathcal{O}(k)$ sobre $\mathbb{C}P^1$:

$$
H_k = H^0(\mathbb{C}P^1, \mathcal{O}(k)) \simeq \mathrm{span}\{1, z, z^2, \ldots, z^k\}, \quad \dim H_k = k+1.
$$

El **kernel de Bergman** asociado es

$$
K_k(z, w) = \sum_{j=0}^{k} s_j(z) \overline{s_j(w)} \propto (1 + z\bar{w})^k.
$$

Con propiedad de reproducción 

$$
f(z) = \int_{\mathbb{C}P^1} K_k(z, w) f(w) \, d\mu(w), \quad \forall f \in H_k.
$$

Este kernel actúa como el análogo matemático del **kernel de Penrose** en la teoría twistorial.

---

## 6. Proyección discreta y método de Nyström

Tomemos un conjunto de $S$ puntos del fibrado, es decir, cada $Z_j \in \mathbb{C}^N$. Se define la matriz de datos

$$
D \in \mathbb{C}^{d \times S}, \quad D[:,j] = |Z_j\rangle, \quad G = D^\dagger D \in \mathbb{C}^{S\times S}, \quad G_{ij}=|Z_i\rangle\langle Z_j|
$$

El operador integral de una proyección continua $K$ actuando en $L^2$ se puede discretizar por puntos de muestreo para dar una aproximación finita-rango. En el entorno finito-dimeno, porque el espacio físico $\mathcal{H}$ ya tiene una dimensión finita $d$, la columna espacio de $D$ es precisamente el subespacio generado por las $S$ secciones muestreadas. La diagonalización de $G$ da la base ortonormal principal asociada al sistema. Concretamente, diagonalizando $G = V \Lambda V^\dagger$, definimos

$$
u_m = \frac{1}{\sqrt{\lambda_m}} D v_m, \quad P = \sum_{m=1}^M |u_m\rangle\langle u_m|.
$$

Entonces $P$ es una proyección ortogonal sobre un subespacio $S \subset \mathcal{H}$ que aproxima el espacio de estados relevantes. En el límite de muestreo denso y con la normalización apropiada converge al proyector de Bergman. 

---

## 7. Hamiltoniano efectivo y teoremas espectrales

El Hamiltoniano proyectado es

$$
H_{\text{eff}} = P H P.
$$

Sea $(\mu, v)$ un autopar de $H_{\text{eff}}$, y definamos el residual $r = (H  - \mu)v$.

### Teorema 1 (error espectral)

$$
\mathrm{dist}(\mu, \mathrm{spec}(H)) \le \|(I - P) H P\|.
$$

### Teorema 2 (Davis–Kahan)

Si $\delta$ es la brecha espectral entre subespacios, entonces

$$
\|\sin\Theta(S, U)\| \le \frac{\|(I - P)HP\|}{\delta}.
$$

---

## 8. Evolución temporal y estimación del error

La evolución exacta y proyectada:

$$
\psi(t) = e^{-iHt}\psi_0, \qquad
\tilde\psi(t) = P e^{-iH_{\text{eff}}t}P\psi_0.
$$

### Teorema 3

Si $\|(I - P) H P\| \le \eta$, entonces

$$
\|\psi(t) - \tilde\psi(t)\| \le \eta\|H\| (e^{\|H\|t} - 1).
$$

---

## 9. Geometría topológica y fibrados en el espacio de estados

El espacio de rayos cuánticos $\mathbb{P}(H)$ es isomorfo a $\mathbb{C}P^{d-1}$.  
Cada vector normalizado $\psi \in H$ define un punto $[\psi] \in \mathbb{C}P^{d-1}$.

La **conexión de Berry** en este fibrado (de fase cuántica) es la conexión natural de $\mathcal{O}(-1)$:

$$
A = i \langle \psi, d\psi \rangle, \qquad F = dA = i\, d\langle \psi, d\psi \rangle.
$$

La curvatura $F$ representa la 2-forma de Kähler de Fubini–Study:

$$
F = \omega_{FS}, \qquad \frac{1}{2\pi}\int_{S^2} F \in \mathbb{Z}.
$$

De esta forma, los estados cuánticos se interpretan como **secciones holomorfas del fibrado tautológico**, y sus fases geométricas (fase de Berry) están gobernadas por la topología compleja de $\mathbb{C}P^{d-1}$.

---

## 10. Interpretación en el contexto del modelo de Ising para su implementación programada.

- Los estados coherentes $|Z\rangle$ forman una parametrización holomorfa del espacio de configuraciones magnéticas.  
- El kernel $(1 + z\bar{w})^k$ genera correlaciones y simetrías $SU(2)$ de rotación de spin.  
- La proyección $P$ corresponde a una **reducción twistorial** del sistema de dimensión $2^N$ a un subespacio holomorfo de dimensión $M$, tal que, $M<2^N$.
- De esta manera, $\mathbb{C}P^n$ provee la estructura geométrica natural para espacios de Hilbert cuánticos, mediante una correspondencia topológica diferencial asociando a cada curva un punto proyectivo.  
- El fibrado tautológico y su métrica de Fubini–Study dan lugar a la conexión de Berry y a la estructura Kähler subyacente, pudiendo evaluar el flujo de Ricci.
- La reducción holomorfa, mediante el kernel de Bergman, permite proyectar Hamiltonianos a subespacios coherentes con error controlado.  
- Es útil y también aplicable a sistemas de baja complejidad de entrelazamiento, como los estados coherentes producto y estaod evolucionando a tiempo cortos discretizados en sistemas sin entrelazamiento masivo.
- En este marco, el modelo de Ising puede reinterpretarse como una dinámica sobre secciones de un fibrado holomorfo, lo que abre el camino a descripciones twistoriales y fibradas de sus soluciones cuánticas, también permitiendo la implementación de la cuantización geométrica y de Moyal no conmutativa.

```python
"""
ising_coherent_projection_enhanced.py

Pipeline mejorado con visualizaciones avanzadas:
 - Gráficas de función de onda y evolución temporal
 - Mapas de calor de matrices y correlaciones
 - Diagramas 3D de la esfera de Bloch y evolución
 - Análisis de fidelidad y error
"""

import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.linalg import svd, eigh, expm
from scipy.sparse.linalg import expm_multiply
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
from matplotlib import cm
from matplotlib.animation import FuncAnimation
from tqdm import tqdm

# =============================================================================
# PARÁMETROS MEJORADOS CON CONTROL DE VISUALIZACIÓN
# =============================================================================

class VisualizationConfig:
    """Configuración centralizada para visualizaciones"""
    def __init__(self):
        self.plot_wavefunction = True
        self.plot_heatmaps = True
        self.plot_3d = True
        self.plot_bloch_sphere = True
        self.plot_correlations = True
        self.save_animations = False
        self.dpi = 150
        self.colormap = 'viridis'

# Parámetros físicos
config = VisualizationConfig()
N = 8                    # número de espines
J = 1.0
h = 0.8
periodic = False

# Muestreo coherente mejorado
S = 300                  
M_trunc = 40            

# Evolución temporal
T = 10.0
nt = 201
t_list = np.linspace(0, T, nt)

# =============================================================================
# FUNCIONES BASE MEJORADAS
# =============================================================================

# Matrices de Pauli (sin cambios)
sx = np.array([[0.0,1.0],[1.0,0.0]], dtype=complex)
sy = np.array([[0.0,-1j],[1j,0.0]], dtype=complex)
sz = np.array([[1.0,0.0],[0.0,-1.0]], dtype=complex)
id2 = np.eye(2, dtype=complex)

def kron_list_dense(mats):
    out = mats[0]
    for A in mats[1:]:
        out = np.kron(out, A)
    return out

def kron_list_sparse(mats):
    out = mats[0]
    for A in mats[1:]:
        out = sp.kron(out, A, format='csr')
    return out

# =============================================================================
# VISUALIZACIÓN DE FUNCIÓN DE ONDA
# =============================================================================

class WavefunctionVisualizer:
    """Visualizador avanzado de funciones de onda"""
    
    @staticmethod
    def plot_wavefunction_amplitude(states, t_list, max_basis_states=50, title="Evolución de la Función de Onda"):
        """
        Gráfica 2D de la amplitud de la función de onda en el tiempo
        """
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
        
        # Limitar estados base para visualización
        n_basis = min(states.shape[1], max_basis_states)
        
        # Amplitud
        amplitude = np.abs(states[:, :n_basis])
        im1 = ax1.imshow(amplitude.T, aspect='auto', origin='lower',
                        extent=[t_list[0], t_list[-1], 0, n_basis],
                        cmap=config.colormap, interpolation='bilinear')
        ax1.set_xlabel('Tiempo')
        ax1.set_ylabel('Estado Base')
        ax1.set_title(f'{title} - Amplitud')
        plt.colorbar(im1, ax=ax1, label='|ψ|')
        
        # Fase
        phase = np.angle(states[:, :n_basis])
        im2 = ax2.imshow(phase.T, aspect='auto', origin='lower',
                        extent=[t_list[0], t_list[-1], 0, n_basis],
                        cmap='twilight', interpolation='bilinear')
        ax2.set_xlabel('Tiempo')
        ax2.set_ylabel('Estado Base')
        ax2.set_title(f'{title} - Fase')
        plt.colorbar(im2, ax=ax2, label='Fase (rad)')
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_probability_density(states, t_list, selected_states=None):
        """
        Densidad de probabilidad para estados seleccionados
        """
        if selected_states is None:
            selected_states = [0, 1, 2, 3]  # Primeros estados base
            
        fig, axes = plt.subplots(2, 2, figsize=(12, 8))
        axes = axes.flatten()
        
        for i, state_idx in enumerate(selected_states):
            if state_idx < states.shape[1]:
                prob = np.abs(states[:, state_idx])**2
                axes[i].plot(t_list, prob, linewidth=2, alpha=0.8)
                axes[i].set_xlabel('Tiempo')
                axes[i].set_ylabel(f'Probabilidad |ψ_{state_idx}|²')
                axes[i].set_title(f'Estado Base {state_idx}')
                axes[i].grid(True, alpha=0.3)
        
        plt.tight_layout()
        return fig

# =============================================================================
# MAPAS DE CALOR MEJORADOS
# =============================================================================

class HeatmapVisualizer:
    """Generador de mapas de calor para matrices y correlaciones"""
    
    @staticmethod
    def plot_hamiltonian_heatmap(H, title="Matriz Hamiltoniana"):
        """
        Mapa de calor de la matriz Hamiltoniana
        """
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
        
        # Parte real
        im1 = ax1.imshow(H.real, cmap='RdBu_r', aspect='auto',
                        norm=colors.CenteredNorm())
        ax1.set_title(f'{title} - Parte Real')
        plt.colorbar(im1, ax=ax1)
        
        # Parte imaginaria
        im2 = ax2.imshow(H.imag, cmap='RdBu_r', aspect='auto',
                        norm=colors.CenteredNorm())
        ax2.set_title(f'{title} - Parte Imaginaria')
        plt.colorbar(im2, ax=ax2)
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_correlation_matrix(states, t_index, title="Matriz de Correlación"):
        """
        Mapa de calor de la matriz de correlación en un tiempo dado
        """
        # Calcular matriz densidad
        psi = states[t_index]
        rho = np.outer(psi, psi.conj())
        
        fig, ax = plt.subplots(figsize=(8, 6))
        im = ax.imshow(np.abs(rho), cmap='hot', aspect='auto',
                      norm=colors.LogNorm(vmin=1e-6, vmax=1.0))
        ax.set_title(f'{title} - t = {t_index}')
        ax.set_xlabel('Estado Base j')
        ax.set_ylabel('Estado Base i')
        plt.colorbar(im, ax=ax, label='|ρ_ij|')
        
        return fig
    
    @staticmethod
    def plot_spin_correlations(spin_ops, states, t_list, N):
        """
        Mapa de calor de correlaciones espín-espín
        """
        n_times = len(t_list)
        corr_matrix = np.zeros((N, N, n_times))
        
        # Calcular correlaciones para cada par de espines
        for i in range(N):
            for j in range(N):
                for t_idx in range(n_times):
                    psi = states[t_idx]
                    # ⟨σ_i^z σ_j^z⟩ - ⟨σ_i^z⟩⟨σ_j^z⟩
                    exp_ij = psi.conj() @ (spin_ops[i] @ (spin_ops[j] @ psi))
                    exp_i = psi.conj() @ (spin_ops[i] @ psi)
                    exp_j = psi.conj() @ (spin_ops[j] @ psi)
                    corr_matrix[i, j, t_idx] = (exp_ij - exp_i * exp_j).real
        
        # Promedio temporal
        corr_avg = np.mean(corr_matrix, axis=2)
        
        fig, ax = plt.subplots(figsize=(8, 6))
        im = ax.imshow(corr_avg, cmap='RdBu_r', aspect='auto',
                      norm=colors.CenteredNorm())
        ax.set_title('Correlaciones Espín-espín Promedio')
        ax.set_xlabel('Espín j')
        ax.set_ylabel('Espín i')
        
        # Añadir valores numéricos
        for i in range(N):
            for j in range(N):
                ax.text(j, i, f'{corr_avg[i,j]:.2f}', 
                       ha="center", va="center", color="black", fontsize=8)
        
        plt.colorbar(im, ax=ax, label='⟨σ_i^z σ_j^z⟩ - ⟨σ_i^z⟩⟨σ_j^z⟩')
        return fig

# =============================================================================
# DIAGRAMAS 3D MEJORADOS
# =============================================================================

class ThreeDVisualizer:
    """Generador de visualizaciones 3D"""
    
    @staticmethod
    def plot_bloch_sphere_3d(z_list, states=None, title="Esfera de Bloch"):
        """
        Visualización 3D de estados en la esfera de Bloch
        """
        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        # Convertir coordenadas z a puntos en la esfera
        theta = 2 * np.arctan(np.abs(z_list))
        phi = np.angle(z_list)
        
        x = np.sin(theta) * np.cos(phi)
        y = np.sin(theta) * np.sin(phi)
        z = np.cos(theta)
        
        # Colorear por amplitud si se proporcionan estados
        if states is not None:
            # Construir un array de colores con la misma longitud que z_list
            colors = None
            states_arr = np.asarray(states)
            if states_arr.ndim == 1 and states_arr.size == len(z_list):
                # Ya corresponde uno a uno con los z
                colors = np.abs(states_arr)
            elif states_arr.ndim == 2:
                # Casos comunes:
                # - filas corresponden a cada z (por ejemplo, vectores por z)
                # - columnas corresponden a cada z
                if states_arr.shape[0] == len(z_list):
                    colors = np.abs(states_arr[:, 0])
                elif states_arr.shape[1] == len(z_list):
                    colors = np.abs(states_arr[0, :])
                else:
                    # No se puede mapear directamente: usar magnitud media como fallback
                    avg = np.mean(np.abs(states_arr))
                    colors = np.full(len(z_list), avg)
            else:
                # Fallback seguro
                colors = np.full(len(z_list), 1.0)
            
            # Asegurar que colors tiene la longitud correcta (recortar o rellenar si es necesario)
            if colors.size != len(z_list):
                colors = np.resize(colors, len(z_list))
            
            scatter = ax.scatter(x, y, z, c=colors, cmap=config.colormap, 
                                 s=50, alpha=0.7)
            plt.colorbar(scatter, ax=ax, label='Amplitud')
        else:
            ax.scatter(x, y, z, s=50, alpha=0.7, color='blue')
        
        # Dibujar esfera
        u = np.linspace(0, 2 * np.pi, 30)
        v = np.linspace(0, np.pi, 30)
        x_sphere = np.outer(np.cos(u), np.sin(v))
        y_sphere = np.outer(np.sin(u), np.sin(v))
        z_sphere = np.outer(np.ones_like(u), np.cos(v))
        
        ax.plot_wireframe(x_sphere, y_sphere, z_sphere, color='gray', 
                         alpha=0.2, linewidth=0.5)
        
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')
        ax.set_title(title)
        
        return fig
    
    @staticmethod
    def plot_evolution_3d(states, t_list, selected_indices=[0, 1, 2]):
        """
        Evolución temporal 3D de amplitudes de estados seleccionados
        """
        fig = plt.figure(figsize=(12, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        for idx in selected_indices:
            if idx < states.shape[1]:
                # Parte real, imaginaria y tiempo
                real_part = states[:, idx].real
                imag_part = states[:, idx].imag
                
                ax.plot(t_list, real_part, imag_part, 
                       label=f'Estado {idx}', linewidth=2, alpha=0.8)
        
        ax.set_xlabel('Tiempo')
        ax.set_ylabel('Parte Real')
        ax.set_zlabel('Parte Imaginaria')
        ax.set_title('Evolución 3D de Estados Cuánticos')
        ax.legend()
        
        return fig
    
    @staticmethod
    def plot_energy_surface_3d(H_eff, parameter_range=(-2, 2), n_points=50):
        """
        Superficie 3D de energía en función de parámetros
        """
        J_vals = np.linspace(parameter_range[0], parameter_range[1], n_points)
        h_vals = np.linspace(parameter_range[0], parameter_range[1], n_points)
        
        J_grid, h_grid = np.meshgrid(J_vals, h_vals)
        energy_grid = np.zeros_like(J_grid)
        
        # Calcular energía base para cada (J, h)
        for i in range(n_points):
            for j in range(n_points):
                # Modificar Hamiltonian con nuevos parámetros (simplificado)
                H_temp = H_eff * (J_grid[i,j] + h_grid[i,j]) / 2.0
                eigvals = np.linalg.eigvalsh(H_temp)
                energy_grid[i,j] = eigvals[0]  # Energía base
        
        fig = plt.figure(figsize=(12, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        surf = ax.plot_surface(J_grid, h_grid, energy_grid, 
                              cmap=config.colormap, alpha=0.8,
                              linewidth=0, antialiased=True)
        
        ax.set_xlabel('J (Acoplamiento)')
        ax.set_ylabel('h (Campo Transversal)')
        ax.set_zlabel('Energía Base')
        ax.set_title('Superficie de Energía Base')
        plt.colorbar(surf, ax=ax, shrink=0.5, aspect=5)
        
        return fig

# =============================================================================
# ANÁLISIS DE FIDELIDAD Y ERROR
# =============================================================================

class FidelityAnalyzer:
    """Análisis cuantitativo de fidelidad y error"""
    
    @staticmethod
    def compute_fidelity_trajectory(states_full, states_proj):
        """
        Calcula fidelidad a lo largo del tiempo
        F(t) = |⟨ψ_full(t)|ψ_proj(t)⟩|²
        """
        fidelity = np.zeros(len(states_full))
        
        for t in range(len(states_full)):
            overlap = np.vdot(states_full[t], states_proj[t])
            fidelity[t] = np.abs(overlap)**2
            
        return fidelity
    
    @staticmethod
    def compute_trace_distance(rho1, rho2):
        """
        Distancia de traza entre matrices densidad
        D(ρ1, ρ2) = 0.5 * Tr|ρ1 - ρ2|
        """
        diff = rho1 - rho2
        eigenvals = np.linalg.eigvals(diff)
        return 0.5 * np.sum(np.abs(eigenvals))
    
    @staticmethod
    def plot_error_analysis(t_list, exp_full, exp_proj, fidelity=None):
        """
        Análisis completo de error y fidelidad
        """
        fig, axes = plt.subplots(2, 2, figsize=(12, 8))
        axes = axes.flatten()
        
        # Error en magnetización
        error_sz = np.abs(exp_full - exp_proj)
        axes[0].plot(t_list, error_sz, 'r-', linewidth=2)
        axes[0].set_xlabel('Tiempo')
        axes[0].set_ylabel('Error |Δ⟨S^z⟩|')
        axes[0].set_title('Error en Magnetización')
        axes[0].grid(True, alpha=0.3)
        axes[0].set_yscale('log')
        
        # Error relativo
        rel_error = error_sz / (np.abs(exp_full) + 1e-12)
        axes[1].plot(t_list, rel_error, 'g-', linewidth=2)
        axes[1].set_xlabel('Tiempo')
        axes[1].set_ylabel('Error Relativo')
        axes[1].set_title('Error Relativo en Magnetización')
        axes[1].grid(True, alpha=0.3)
        axes[1].set_yscale('log')
        
        # Fidelidad
        if fidelity is not None:
            axes[2].plot(t_list, fidelity, 'b-', linewidth=2)
            axes[2].set_xlabel('Tiempo')
            axes[2].set_ylabel('Fidelidad')
            axes[2].set_title('Fidelidad entre Estados')
            axes[2].grid(True, alpha=0.3)
            axes[2].set_ylim(0, 1.1)
        
        # Error acumulado
        cumulative_error = np.cumsum(error_sz) * (t_list[1] - t_list[0])
        axes[3].plot(t_list, cumulative_error, 'm-', linewidth=2)
        axes[3].set_xlabel('Tiempo')
        axes[3].set_ylabel('Error Acumulado')
        axes[3].set_title('Error Acumulado en el Tiempo')
        axes[3].grid(True, alpha=0.3)
        
        plt.tight_layout()
        return fig

# =============================================================================
# FUNCIONES DEL SISTEMA ISING (CON MEJORAS)
# =============================================================================

def ising_hamiltonian_sparse(N, J=1.0, h=0.5, periodic=False):
    """Hamiltoniano de Ising mejorado con más información"""
    dim = 2**N
    H = sp.csr_matrix((dim, dim), dtype=complex)
    
    # Interacción ZZ
    for i in range(N-1 + int(periodic)):
        j = (i+1) % N
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sz)
        ops[j] = sp.csr_matrix(sz)
        H = H - J * kron_list_sparse(ops)
    
    # Campo transversal
    for i in range(N):
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sx)
        H = H - h * kron_list_sparse(ops)
    
    print(f"Hamiltoniano: N={N}, J={J}, h={h}, nnz={H.nnz}/{dim*dim}")
    return H

def sample_cp1_uniform(S):
    """Muestreo mejorado de la esfera de Bloch"""
    u = np.random.uniform(-1.0, 1.0, size=S)
    phi = np.random.uniform(0.0, 2*np.pi, size=S)
    theta = np.arccos(u)
    z = np.tan(theta/2.0) * np.exp(1j*phi)
    return z, theta, phi

def build_spin_operators(N):
    """Construye operadores de spin individuales"""
    spin_ops = []
    for i in range(N):
        ops = [id2] * N
        ops[i] = sz
        spin_mat = kron_list_dense(ops)
        spin_ops.append(spin_mat)
    return spin_ops

# =============================================================================
# FLUJO PRINCIPAL MEJORADO
# =============================================================================

def main_enhanced():
    """Flujo principal con visualizaciones mejoradas"""
    
    print("=== SISTEMA ISING CON PROYECCIÓN COHERENTE MEJORADO ===")
    
    # Construir Hamiltonianos y operadores
    H = ising_hamiltonian_sparse(N, J=J, h=h, periodic=periodic)
    dim = 2**N
    
    # Muestreo coherente mejorado
    z_list, theta, phi = sample_cp1_uniform(S)
    Data = build_product_coherent_vectors(N, z_list)
    norms = np.linalg.norm(Data, axis=0)
    Data = Data / norms[np.newaxis, :]
    
    # Construir proyección
    if dim > S:
        print("Usando descomposición de Gram...")
        P, evals = build_projection_from_gram(Data, M_trunc)
    else:
        print("Usando SVD directo...")
        P, svals = build_projection_from_data_matrix(Data, M_trunc)
    
    M = P.shape[1]
    print(f"Proyección: dim={dim} → M={M}")
    
    # Estado inicial y operadores
    psi0 = np.zeros((dim,), dtype=complex)
    psi0[0] = 1.0
    Sz_total = build_Sz_total_sparse(N)
    spin_ops = build_spin_operators(N)
    
    # Evolución proyectada
    psi_proj_lifted, H_eff, psi0_proj = project_and_evolve(H, P, psi0, t_list)
    
    # Evolución completa (si es posible)
    print("Calculando evolución completa...")
    try:
        A = -1j * H
        psi_full_t = expm_multiply(A, psi0, start=0.0, stop=T, num=len(t_list))
        full_available = True
    except Exception as e:
        print(f"Evolución completa no disponible: {e}")
        psi_full_t = None
        full_available = False
    
    # Cálculo de observables
    exp_proj = np.empty(len(t_list), dtype=float)
    exp_full = np.empty(len(t_list), dtype=float) if full_available else None
    
    for k in tqdm(range(len(t_list)), desc="Calculando observables"):
        v = psi_proj_lifted[k]
        exp_proj[k] = (v.conj() @ (Sz_total @ v)).real / N
        if full_available:
            u = psi_full_t[k]
            exp_full[k] = (u.conj() @ (Sz_total @ u)).real / N
    
    # =========================================================================
    # VISUALIZACIONES MEJORADAS
    # =========================================================================
    
    print("\n=== GENERANDO VISUALIZACIONES ===")
    
    # 1. Función de onda
    if config.plot_wavefunction:
        print("Generando gráficas de función de onda...")
        fig_wave = WavefunctionVisualizer.plot_wavefunction_amplitude(
            psi_proj_lifted, t_list, title="Función de Onda Proyectada")
        plt.savefig('wavefunction_evolution.png', dpi=config.dpi, bbox_inches='tight')
        
        fig_prob = WavefunctionVisualizer.plot_probability_density(
            psi_proj_lifted, t_list)
        plt.savefig('probability_density.png', dpi=config.dpi, bbox_inches='tight')
    
    # 2. Mapas de calor
    if config.plot_heatmaps:
        print("Generando mapas de calor...")
        fig_hamiltonian = HeatmapVisualizer.plot_hamiltonian_heatmap(
            H_eff, title="Hamiltoniano Efectivo")
        plt.savefig('hamiltonian_heatmap.png', dpi=config.dpi, bbox_inches='tight')
        
        fig_correlation = HeatmapVisualizer.plot_correlation_matrix(
            psi_proj_lifted, len(t_list)//2)
        plt.savefig('correlation_matrix.png', dpi=config.dpi, bbox_inches='tight')
        
        if config.plot_correlations:
            fig_spin_corr = HeatmapVisualizer.plot_spin_correlations(
                spin_ops, psi_proj_lifted, t_list, N)
            plt.savefig('spin_correlations.png', dpi=config.dpi, bbox_inches='tight')
    
    # 3. Visualizaciones 3D
    if config.plot_3d:
        print("Generando visualizaciones 3D...")
        fig_bloch = ThreeDVisualizer.plot_bloch_sphere_3d(
            z_list, psi_proj_lifted, "Estados Coherentes en Esfera de Bloch")
        plt.savefig('bloch_sphere.png', dpi=config.dpi, bbox_inches='tight')
        
        fig_evolution_3d = ThreeDVisualizer.plot_evolution_3d(
            psi_proj_lifted, t_list)
        plt.savefig('evolution_3d.png', dpi=config.dpi, bbox_inches='tight')
        
        fig_energy = ThreeDVisualizer.plot_energy_surface_3d(H_eff)
        plt.savefig('energy_surface.png', dpi=config.dpi, bbox_inches='tight')
    
    # 4. Análisis de error y fidelidad
    if full_available:
        print("Analizando fidelidad y error...")
        fidelity = FidelityAnalyzer.compute_fidelity_trajectory(
            psi_full_t, psi_proj_lifted)
        
        fig_error = FidelityAnalyzer.plot_error_analysis(
            t_list, exp_full, exp_proj, fidelity)
        plt.savefig('error_analysis.png', dpi=config.dpi, bbox_inches='tight')
        
        # Métricas finales
        avg_fidelity = np.mean(fidelity)
        max_error = np.max(np.abs(exp_full - exp_proj))
        print(f"\n=== MÉTRICAS FINALES ===")
        print(f"Fidelidad promedio: {avg_fidelity:.6f}")
        print(f"Error máximo en ⟨S^z⟩: {max_error:.6f}")
        print(f"Dimensión reducida: {M}/{dim} ({M/dim*100:.1f}%)")
    
    # 5. Comparación clásica
    plt.figure(figsize=(10, 6))
    plt.plot(t_list, exp_proj, 'b-', linewidth=2, label='Proyectado', alpha=0.8)
    if full_available:
        plt.plot(t_list, exp_full, 'r--', linewidth=2, label='Completo', alpha=0.8)
    plt.xlabel('Tiempo')
    plt.ylabel(r'$\langle S^z \rangle$ por sitio')
    plt.title(f'Comparación de Evolución (N={N}, M={M})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.savefig('comparison_final.png', dpi=config.dpi, bbox_inches='tight')
    
    plt.show()
    
    print("\n=== ANÁLISIS COMPLETADO ===")
    return psi_proj_lifted, H_eff, P

# Las funciones auxiliares restantes se mantienen igual que en tu código original
# (build_product_coherent_vectors, build_projection_from_data_matrix, 
# build_projection_from_gram, project_and_evolve, build_Sz_total_sparse)

if __name__ == "__main__":
    # Ejecutar análisis completo
    results = main_enhanced()
```

```python
"""
ising_coherent_projection_complete.py
Implementación completa del modelo de Ising con:
- Proyección holomorfa mediante estados coherentes
- Flujo de Ricci geométrico con parámetro β
- Evolución temporal completa de la función de onda
- Análisis espectral y de fidelidad
- Visualizaciones avanzadas de la dinámica cuántica
"""

import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.linalg import svd, eigh, expm
from scipy.sparse.linalg import expm_multiply
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
from matplotlib import cm
from matplotlib.animation import FuncAnimation
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# PARÁMETROS DEL SISTEMA Y CONFIGURACIÓN
# =============================================================================

class SystemConfig:
    """Configuración completa del sistema físico"""
    def __init__(self):
        # Parámetros físicos
        self.N = 6                    # número de espines (reducido para demo)
        self.J = 1.0                  # acoplamiento
        self.h = 0.5                  # campo transversal
        self.periodic = False
        
        # Parámetros de muestreo coherente
        self.S = 200                  # número de puntos de muestreo
        self.M_trunc = 30             # dimensión del subespacio proyectado
        
        # Parámetros de evolución temporal
        self.T = 8.0                  # tiempo total
        self.nt = 201                 # puntos temporales
        self.t_list = np.linspace(0, self.T, self.nt)
        
        # Parámetros del flujo de Ricci
        self.beta_list = np.logspace(-3, 2, 20)  # rangos de β
        self.use_ricci_flow = True    # activar flujo geométrico
        
        # Configuración de visualización
        self.plot_wavefunction = True
        self.plot_ricci_flow = True
        self.plot_3d = True
        self.plot_spectral = True
        self.dpi = 150
        self.colormap = 'viridis'

# =============================================================================
# FUNCIONES MATEMÁTICAS BASE - IMPLEMENTACIÓN MEJORADA
# =============================================================================

# Matrices de Pauli
sx = np.array([[0.0, 1.0], [1.0, 0.0]], dtype=complex)
sy = np.array([[0.0, -1j], [1j, 0.0]], dtype=complex)
sz = np.array([[1.0, 0.0], [0.0, -1.0]], dtype=complex)
id2 = np.eye(2, dtype=complex)

def kron_list_dense(mats):
    """Producto de Kronecker para lista de matrices"""
    out = mats[0]
    for A in mats[1:]:
        out = np.kron(out, A)
    return out

def kron_list_sparse(mats):
    """Producto de Kronecker sparse para lista de matrices"""
    out = mats[0]
    for A in mats[1:]:
        out = sp.kron(out, A, format='csr')
    return out

def coherent_state(z):
    """Estado coherente para un solo espín en CP^1"""
    norm = np.sqrt(1 + np.abs(z)**2)
    return np.array([1.0, z], dtype=complex) / norm

def build_product_coherent_vectors(N, z_list):
    """
    Construye estados coherentes producto para N espines
    |Z⟩ = ⊗_{i=1}^N |z_i⟩
    """
    S = len(z_list)
    dim = 2**N
    Data = np.zeros((dim, S), dtype=complex)
    
    for j, z in enumerate(z_list):
        # Estado coherente para cada espín
        single_spin_states = [coherent_state(z) for _ in range(N)]
        
        # Construir estado producto
        psi = single_spin_states[0]
        for i in range(1, N):
            psi = np.kron(psi, single_spin_states[i])
        
        Data[:, j] = psi
    
    return Data

def build_projection_from_data_matrix(Data, M):
    """Construye proyección usando SVD de la matriz de datos"""
    U, s, Vh = svd(Data, full_matrices=False)
    P = U[:, :M]
    return P, s[:M]

def build_projection_from_gram(Data, M):
    """Construye proyección usando diagonalización de Gram para sistemas grandes"""
    G = Data.conj().T @ Data
    evals, evecs = eigh(G)
    # Ordenar eigenvalues descendentes
    idx = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:, idx]
    
    # Construir base ortonormal
    V = evecs[:, :M]
    Lambda_sqrt = np.sqrt(evals[:M])
    P = Data @ (V / Lambda_sqrt[np.newaxis, :])
    
    return P, evals[:M]

def ising_hamiltonian_sparse(N, J=1.0, h=0.5, periodic=False):
    """Hamiltoniano de Ising con interacciones ZZ y campo transversal"""
    dim = 2**N
    H = sp.csr_matrix((dim, dim), dtype=complex)
    
    # Interacción ZZ
    for i in range(N-1 + int(periodic)):
        j = (i+1) % N
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sz)
        ops[j] = sp.csr_matrix(sz)
        H = H - J * kron_list_sparse(ops)
    
    # Campo transversal
    for i in range(N):
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sx)
        H = H - h * kron_list_sparse(ops)
    
    print(f"Hamiltoniano: N={N}, J={J}, h={h}, nnz={H.nnz}/{dim*dim}")
    return H

def build_Sz_total_sparse(N):
    """Operador total S^z = Σ_i σ_i^z"""
    dim = 2**N
    Sz_total = sp.csr_matrix((dim, dim), dtype=complex)
    
    for i in range(N):
        ops = [id2] * N
        ops[i] = sz
        Sz_total = Sz_total + kron_list_sparse([sp.csr_matrix(op) for op in ops])
    
    return Sz_total

def sample_cp1_uniform(S):
    """Muestreo uniforme mejorado de CP^1 (esfera de Bloch)"""
    u = np.random.uniform(-1.0, 1.0, size=S)
    phi = np.random.uniform(0.0, 2*np.pi, size=S)
    theta = np.arccos(u)
    z = np.tan(theta/2.0) * np.exp(1j*phi)
    return z, theta, phi

def sample_cp1_fibonacci(S):
    """Muestreo quasi-uniforme usando secuencia de Fibonacci"""
    indices = np.arange(0, S, dtype=float)
    phi = 2 * np.pi * indices * (1.0 / ((1.0 + np.sqrt(5)) / 2.0))
    theta = np.arccos(1.0 - 2.0 * (indices + 0.5) / S)
    z = np.tan(theta/2.0) * np.exp(1j*phi)
    return z, theta, phi

# =============================================================================
# FLUJO DE RICCI Y PROYECCIÓN PONDERADA
# =============================================================================

class RicciFlowProjector:
    """Implementa el flujo de Ricci geométrico mediante ponderación de kernel"""
    
    def __init__(self, H, beta_list):
        self.H = H
        self.beta_list = beta_list
        self.dim = H.shape[0]
        
    def compute_weighted_kernel(self, Data, beta):
        """Calcula kernel ponderado con pesos e^{-βE}"""
        S = Data.shape[1]
        weights = np.zeros(S, dtype=float)
        
        for j in range(S):
            z_state = Data[:, j]
            # ⟨Z|e^{-βH}|Z⟩ ≈ e^{-β⟨Z|H|Z⟩} (aproximación de punto silla)
            energy_exp = np.vdot(z_state, self.H @ z_state).real
            weights[j] = np.exp(-beta * energy_exp)
        
        # Normalizar pesos
        weights = weights / np.sum(weights)
        return weights
    
    def build_ricci_projection(self, Data, M, beta):
        """Construye proyección con métrica vestida por flujo de Ricci"""
        weights = self.compute_weighted_kernel(Data, beta)
        
        # Matriz de datos ponderada
        W_sqrt = np.sqrt(weights)
        Data_weighted = Data * W_sqrt[np.newaxis, :]
        
        # Proyección SVD sobre datos ponderados
        if self.dim > Data.shape[1]:
            P, evals = build_projection_from_gram(Data_weighted, M)
        else:
            P, svals = build_projection_from_data_matrix(Data_weighted, M)
            
        return P, weights
    
    def compute_fidelity_curve(self, Data, psi0, M):
        """Calcula curva de fidelidad F(β) = |⟨ψ0|P(β)P(β)^†|ψ0⟩|^2"""
        fidelity_list = []
        
        for beta in tqdm(self.beta_list, desc="Flujo de Ricci"):
            P, weights = self.build_ricci_projection(Data, M, beta)
            # Proyector al subespacio
            Proj = P @ P.conj().T
            # Fidelidad
            fid = np.abs(np.vdot(psi0, Proj @ psi0))**2
            fidelity_list.append(fid)
            
        return np.array(fidelity_list)

# =============================================================================
# EVOLUCIÓN TEMPORAL Y DINÁMICA CUÁNTICA
# =============================================================================

def project_and_evolve(H, P, psi0, t_list):
    """Proyecta Hamiltonian y evoluciona estado inicial"""
    # Hamiltoniano efectivo
    H_eff = P.conj().T @ (H @ P)
    
    # Estado inicial proyectado
    psi0_proj = P.conj().T @ psi0
    norm0 = np.linalg.norm(psi0_proj)
    psi0_proj = psi0_proj / norm0
    
    # Evolución temporal
    psi_proj_t = []
    for t in t_list:
        U_eff = expm(-1j * H_eff * t)
        psi_t = U_eff @ psi0_proj
        psi_proj_t.append(psi_t)
    
    # Levantar estados al espacio completo
    psi_proj_lifted = [P @ psi for psi in psi_proj_t]
    
    return np.array(psi_proj_lifted), H_eff, psi0_proj

def compute_observables(states, operators):
    """Calcula valores esperados para una lista de operadores"""
    expectations = []
    for op in operators:
        exp_vals = [np.vdot(psi, op @ psi).real for psi in states]
        expectations.append(exp_vals)
    return np.array(expectations)

def compute_entanglement_entropy(states, N, subsystem_size=1):
    """Calcula entropía de entrelazamiento para subsistema"""
    entropies = []
    dim_A = 2**subsystem_size
    dim_B = 2**(N - subsystem_size)
    
    for psi in states:
        # Reorganizar en matriz para traza parcial
        psi_matrix = psi.reshape(dim_A, dim_B)
        # Matriz densidad reducida
        rho_A = psi_matrix @ psi_matrix.conj().T
        # Eigenvalores
        eigenvals = np.linalg.eigvalsh(rho_A)
        eigenvals = eigenvals[eigenvals > 1e-12]  # evitar log(0)
        # Entropía de von Neumann
        S = -np.sum(eigenvals * np.log(eigenvals))
        entropies.append(S)
    
    return np.array(entropies)

# =============================================================================
# VISUALIZACIONES AVANZADAS
# =============================================================================

class QuantumVisualizer:
    """Visualizador completo de la dinámica cuántica"""
    
    @staticmethod
    def plot_ricci_flow(beta_list, fidelity_list, optimal_beta=None):
        """Gráfica del flujo de Ricci: fidelidad vs β"""
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.semilogx(beta_list, fidelity_list, 'bo-', linewidth=2, markersize=6)
        ax.set_xlabel('Parámetro β (Escala Geométrica)')
        ax.set_ylabel('Fidelidad F(β)')
        ax.set_title('Flujo de Ricci: Fidelidad vs Escala Geométrica')
        ax.grid(True, alpha=0.3)
        
        if optimal_beta is not None:
            ax.axvline(optimal_beta, color='red', linestyle='--', 
                      label=f'β óptimo = {optimal_beta:.3f}')
            ax.legend()
        
        ax.set_ylim(0, 1.1)
        return fig
    
    @staticmethod
    def plot_wavefunction_dynamics(states, t_list, title="Evolución de la Función de Onda"):
        """Visualización 2D de la amplitud y fase de la función de onda"""
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
        
        # Amplitud
        amplitude = np.abs(states)
        im1 = ax1.imshow(amplitude.T, aspect='auto', origin='lower',
                        extent=[t_list[0], t_list[-1], 0, states.shape[1]],
                        cmap='viridis', interpolation='bilinear')
        ax1.set_ylabel('Estado Base')
        ax1.set_title(f'{title} - Amplitud')
        plt.colorbar(im1, ax=ax1, label='|ψ|')
        
        # Fase
        phase = np.angle(states)
        im2 = ax2.imshow(phase.T, aspect='auto', origin='lower',
                        extent=[t_list[0], t_list[-1], 0, states.shape[1]],
                        cmap='twilight', interpolation='bilinear')
        ax2.set_xlabel('Tiempo')
        ax2.set_ylabel('Estado Base')
        ax2.set_title(f'{title} - Fase')
        plt.colorbar(im2, ax=ax2, label='Fase (rad)')
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_observables_dynamics(t_list, observables_dict):
        """Gráfica de múltiples observables en el tiempo"""
        fig, axes = plt.subplots(2, 2, figsize=(12, 8))
        axes = axes.flatten()
        
        observables_items = list(observables_dict.items())
        for i in range(min(4, len(observables_items))):
            name, data = observables_items[i]
            axes[i].plot(t_list, data, linewidth=2)
            axes[i].set_xlabel('Tiempo')
            axes[i].set_ylabel(name)
            axes[i].set_title(f'Evolución de {name}')
            axes[i].grid(True, alpha=0.3)
        
        # Ocultar ejes no utilizados
        for i in range(len(observables_items), 4):
            axes[i].set_visible(False)
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_spectral_properties(H, H_eff, P):
        """Comparación de propiedades espectrales"""
        try:
            # Espectro completo (primeros 50 estados)
            k = min(50, H.shape[0]-2)
            eigvals_full = spla.eigsh(H, k=k, return_eigenvectors=False, which='SR')
            
            # Espectro efectivo
            eigvals_eff = eigh(H_eff)[0]
            
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
            
            # Densidad espectral
            ax1.hist(eigvals_full, bins=30, alpha=0.7, label='Completo', density=True)
            ax1.hist(eigvals_eff, bins=20, alpha=0.7, label='Proyectado', density=True)
            ax1.set_xlabel('Energía')
            ax1.set_ylabel('Densidad')
            ax1.set_title('Densidad Espectral')
            ax1.legend()
            
            # Error espectral (solo para los primeros M estados)
            n_eff = min(len(eigvals_eff), len(eigvals_full))
            spectral_error = np.abs(eigvals_eff[:n_eff] - eigvals_full[:n_eff])
            ax2.semilogy(range(n_eff), spectral_error, 'ro-')
            ax2.set_xlabel('Índice del Estado')
            ax2.set_ylabel('Error Espectral')
            ax2.set_title('Error en Autovalores')
            ax2.grid(True, alpha=0.3)
            
            plt.tight_layout()
            return fig
        except Exception as e:
            print(f"Error en plot_spectral_properties: {e}")
            return None
    
    @staticmethod
    def plot_bloch_evolution_3d(z_list, states, t_list, N, n_snapshots=5):
        """Evolución 3D en la esfera de Bloch"""
        try:
            fig = plt.figure(figsize=(12, 8))
            ax = fig.add_subplot(111, projection='3d')
            
            # Seleccionar snapshots temporales
            time_indices = np.linspace(0, len(t_list)-1, n_snapshots, dtype=int)
            colors = cm.viridis(np.linspace(0, 1, n_snapshots))
            
            for idx, color in zip(time_indices, colors):
                # Estados en tiempo t
                psi_t = states[idx]
                
                # Calcular amplitudes en base coherente
                amplitudes = []
                for z in z_list:
                    # Estado coherente |z⟩ para un espín
                    coh_state_single = coherent_state(z)
                    # Producto tensorial para N espines
                    coh_state_full = coh_state_single
                    for _ in range(N-1):
                        coh_state_full = np.kron(coh_state_full, coh_state_single)
                    
                    # Amplitud de superposición
                    amp = np.vdot(coh_state_full, psi_t)
                    amplitudes.append(amp)
                
                amplitudes = np.array(amplitudes)
                probabilities = np.abs(amplitudes)**2
                
                # Coordenadas en esfera de Bloch
                theta = 2 * np.arctan(np.abs(z_list))
                phi = np.angle(z_list)
                x = np.sin(theta) * np.cos(phi)
                y = np.sin(theta) * np.sin(phi)
                z_coord = np.cos(theta)
                
                ax.scatter(x, y, z_coord, c=[color], s=50*probabilities, 
                          alpha=0.6, label=f't={t_list[idx]:.1f}')
            
            # Esfera de referencia
            u = np.linspace(0, 2 * np.pi, 20)
            v = np.linspace(0, np.pi, 20)
            x_sphere = np.outer(np.cos(u), np.sin(v))
            y_sphere = np.outer(np.sin(u), np.sin(v))
            z_sphere = np.outer(np.ones_like(u), np.cos(v))
            ax.plot_wireframe(x_sphere, y_sphere, z_sphere, color='gray', 
                             alpha=0.1, linewidth=0.5)
            
            ax.set_xlabel('X')
            ax.set_ylabel('Y')
            ax.set_zlabel('Z')
            ax.set_title('Evolución en Esfera de Bloch')
            ax.legend()
            
            return fig
        except Exception as e:
            print(f"Error en plot_bloch_evolution_3d: {e}")
            return None

# =============================================================================
# ANÁLISIS DE ERROR Y FIDELIDAD
# =============================================================================

class ErrorAnalyzer:
    """Análisis cuantitativo de errores de aproximación"""
    
    @staticmethod
    def compute_fidelity_trajectory(states_full, states_proj):
        """Calcula fidelidad F(t) = |⟨ψ_full(t)|ψ_proj(t)⟩|²"""
        fidelity = np.zeros(len(states_full))
        for t in range(len(states_full)):
            overlap = np.vdot(states_full[t], states_proj[t])
            fidelity[t] = np.abs(overlap)**2
        return fidelity
    
    @staticmethod
    def compute_observable_errors(obs_full, obs_proj):
        """Calcula errores en observables"""
        absolute_error = np.abs(obs_full - obs_proj)
        relative_error = absolute_error / (np.abs(obs_full) + 1e-12)
        return absolute_error, relative_error

# =============================================================================
# FLUJO PRINCIPAL COMPLETO
# =============================================================================

def main_complete():
    """Implementación completa del sistema Ising con proyección holomorfa"""
    config = SystemConfig()
    print("=== SISTEMA ISING CON PROYECCIÓN HOLOMORFA COMPLETA ===")
    print(f"N = {config.N}, J = {config.J}, h = {config.h}")
    print(f"Dimensión completa: 2^{config.N} = {2**config.N}")
    
    # 1. CONSTRUCCIÓN DEL SISTEMA FÍSICO
    print("\n1. Construyendo Hamiltonianos...")
    H = ising_hamiltonian_sparse(config.N, config.J, config.h, config.periodic)
    Sz_total = build_Sz_total_sparse(config.N)
    dim = 2**config.N
    
    # Estado inicial |00...0⟩
    psi0 = np.zeros(dim, dtype=complex)
    psi0[0] = 1.0
    
    # 2. MUESTREO COHERENTE Y FLUJO DE RICCI
    print("\n2. Muestreo coherente y flujo de Ricci...")
    z_list, theta, phi = sample_cp1_fibonacci(config.S)
    Data = build_product_coherent_vectors(config.N, z_list)
    
    # Normalizar estados coherentes
    norms = np.linalg.norm(Data, axis=0)
    Data = Data / norms[np.newaxis, :]
    
    # Optimizar β usando flujo de Ricci
    if config.use_ricci_flow:
        ricci_projector = RicciFlowProjector(H, config.beta_list)
        fidelity_curve = ricci_projector.compute_fidelity_curve(Data, psi0, config.M_trunc)
        
        # Encontrar β óptimo
        optimal_idx = np.argmax(fidelity_curve)
        optimal_beta = config.beta_list[optimal_idx]
        print(f"β óptimo: {optimal_beta:.4f} (F = {fidelity_curve[optimal_idx]:.6f})")
        
        # Construir proyección con β óptimo
        P, weights = ricci_projector.build_ricci_projection(Data, config.M_trunc, optimal_beta)
    else:
        # Proyección estándar (β = 0)
        if dim > config.S:
            P, evals = build_projection_from_gram(Data, config.M_trunc)
        else:
            P, svals = build_projection_from_data_matrix(Data, config.M_trunc)
        optimal_beta = 0.0
        fidelity_curve = None
    
    M = P.shape[1]
    print(f"Dimensión reducida: {M}/{dim} ({M/dim*100:.2f}%)")
    
    # 3. EVOLUCIÓN TEMPORAL
    print("\n3. Evolución temporal...")
    
    # Evolución proyectada
    psi_proj_lifted, H_eff, psi0_proj = project_and_evolve(H, P, psi0, config.t_list)
    
    # Evolución completa (para comparación)
    print("Calculando evolución completa...")
    try:
        if dim <= 4096:  # Límite para cálculo completo
            psi_full_t = []
            for t in config.t_list:
                U = expm(-1j * H * t)
                psi_t = U @ psi0
                psi_full_t.append(psi_t)
            psi_full_t = np.array(psi_full_t)
            full_available = True
        else:
            print("Dimensión demasiado grande para evolución completa")
            psi_full_t = None
            full_available = False
    except Exception as e:
        print(f"Evolución completa no disponible: {e}")
        psi_full_t = None
        full_available = False
    
    # 4. CÁLCULO DE OBSERVABLES
    print("\n4. Calculando observables...")
    
    # Magnetización
    exp_sz_proj = np.array([np.vdot(psi, Sz_total @ psi).real / config.N 
                           for psi in psi_proj_lifted])
    
    if full_available:
        exp_sz_full = np.array([np.vdot(psi, Sz_total @ psi).real / config.N 
                               for psi in psi_full_t])
    
    # Entropía de entrelazamiento
    entropy_proj = compute_entanglement_entropy(psi_proj_lifted, config.N, 1)
    
    # 5. ANÁLISIS DE ERROR
    print("\n5. Análisis de error...")
    
    if full_available:
        # Fidelidad
        fidelity = ErrorAnalyzer.compute_fidelity_trajectory(psi_full_t, psi_proj_lifted)
        avg_fidelity = np.mean(fidelity)
        
        # Error en observables
        abs_error, rel_error = ErrorAnalyzer.compute_observable_errors(exp_sz_full, exp_sz_proj)
        max_abs_error = np.max(abs_error)
        
        print(f"Fidelidad promedio: {avg_fidelity:.6f}")
        print(f"Error máximo en ⟨S^z⟩: {max_abs_error:.6f}")
    else:
        fidelity = None
        avg_fidelity = None
        max_abs_error = None
    
    # 6. VISUALIZACIONES
    print("\n6. Generando visualizaciones...")
    
    visualizer = QuantumVisualizer()
    
    # Flujo de Ricci
    if config.plot_ricci_flow and fidelity_curve is not None:
        fig_ricci = visualizer.plot_ricci_flow(config.beta_list, fidelity_curve, optimal_beta)
        plt.savefig('ricci_flow.png', dpi=config.dpi, bbox_inches='tight')
        plt.show()
    
    # Dinámica de función de onda
    if config.plot_wavefunction:
        fig_wave = visualizer.plot_wavefunction_dynamics(psi_proj_lifted, config.t_list)
        plt.savefig('wavefunction_dynamics.png', dpi=config.dpi, bbox_inches='tight')
        plt.show()
    
    # Observables
    observables_dict = {
        '⟨S^z⟩': exp_sz_proj,
        'Entropía': entropy_proj
    }
    if full_available and fidelity is not None:
        observables_dict['Fidelidad'] = fidelity
        
    fig_obs = visualizer.plot_observables_dynamics(config.t_list, observables_dict)
    plt.savefig('observables_dynamics.png', dpi=config.dpi, bbox_inches='tight')
    plt.show()
    
    # Propiedades espectrales
    if config.plot_spectral:
        fig_spec = visualizer.plot_spectral_properties(H, H_eff, P)
        if fig_spec is not None:
            plt.savefig('spectral_properties.png', dpi=config.dpi, bbox_inches='tight')
            plt.show()
    
    # Evolución 3D
    if config.plot_3d:
        fig_3d = visualizer.plot_bloch_evolution_3d(z_list[:20], psi_proj_lifted, config.t_list, config.N)
        if fig_3d is not None:
            plt.savefig('bloch_evolution_3d.png', dpi=config.dpi, bbox_inches='tight')
            plt.show()
    
    # 7. COMPARACIÓN FINAL
    plt.figure(figsize=(12, 8))
    
    plt.subplot(2, 2, 1)
    plt.plot(config.t_list, exp_sz_proj, 'b-', linewidth=2, label='Proyectado')
    if full_available:
        plt.plot(config.t_list, exp_sz_full, 'r--', linewidth=2, label='Completo')
    plt.xlabel('Tiempo')
    plt.ylabel(r'$\langle S^z \rangle$')
    plt.title('Magnetización')
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.subplot(2, 2, 2)
    plt.plot(config.t_list, entropy_proj, 'g-', linewidth=2)
    plt.xlabel('Tiempo')
    plt.ylabel('Entropía')
    plt.title('Entropía de Entrelazamiento')
    plt.grid(True, alpha=0.3)
    
    if full_available and fidelity is not None:
        plt.subplot(2, 2, 3)
        plt.plot(config.t_list, fidelity, 'm-', linewidth=2)
        plt.xlabel('Tiempo')
        plt.ylabel('Fidelidad')
        plt.title('Fidelidad')
        plt.grid(True, alpha=0.3)
        plt.ylim(0, 1.1)
        
        plt.subplot(2, 2, 4)
        plt.semilogy(config.t_list, abs_error, 'r-', linewidth=2)
        plt.xlabel('Tiempo')
        plt.ylabel('Error Absoluto')
        plt.title('Error en ⟨S^z⟩')
        plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('final_comparison.png', dpi=config.dpi, bbox_inches='tight')
    plt.show()
    
    # 8. MÉTRICAS FINALES
    print("\n=== MÉTRICAS FINALES ===")
    print(f"Dimensión reducida: {M}/{dim} ({M/dim*100:.2f}%)")
    if full_available:
        print(f"Fidelidad promedio: {avg_fidelity:.6f}")
        print(f"Error máximo en ⟨S^z⟩: {max_abs_error:.6f}")
    print(f"β óptimo: {optimal_beta:.4f}")
    
    return {
        'H_eff': H_eff,
        'P': P,
        'psi_proj': psi_proj_lifted,
        'fidelity_curve': fidelity_curve,
        'optimal_beta': optimal_beta,
        'config': config
    }

# =============================================================================
# EJECUCIÓN
# =============================================================================

if __name__ == "__main__":
    results = main_complete()
    print("\n=== ANÁLISIS COMPLETADO ===")
```

```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.linalg import eigh
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

class ScalableCoherentProjection:
    """
    Implementación escalable corregida de la proyección coherente
    """
    
    def __init__(self, N, J=1.0, h=0.8, S=500, M_trunc=50):
        self.N = N
        self.J = J
        self.h = h
        self.S = S
        self.M_trunc = M_trunc
        self.dim = 2**N
        
        # Operadores de Pauli
        self.sx = np.array([[0, 1], [1, 0]], dtype=complex)
        self.sz = np.array([[1, 0], [0, -1]], dtype=complex)
        self.id2 = np.eye(2, dtype=complex)
        
    def build_ising_hamiltonian_sparse_optimized(self):
        """Hamiltoniano de Ising optimizado"""
        print(f"Construyendo Hamiltoniano para N={self.N}...")
        
        H = sp.csr_matrix((self.dim, self.dim), dtype=complex)
        
        # Construcción eficiente de términos ZZ
        for i in range(self.N - 1):
            # Producto tensorial manual para evitar grandes matrices temporales
            left_dims = 2**i
            middle_dims = 2**(self.N - i - 2)
            
            # Construir término ZZ para posiciones i, i+1
            sz_sz = sp.kron(self.sz, self.sz)
            if left_dims > 1:
                left = sp.eye(left_dims, dtype=complex)
                sz_sz = sp.kron(left, sz_sz)
            if middle_dims > 1:
                right = sp.eye(middle_dims, dtype=complex)
                sz_sz = sp.kron(sz_sz, right)
                
            H -= self.J * sz_sz
        
        # Campo transversal
        for i in range(self.N):
            left_dims = 2**i
            right_dims = 2**(self.N - i - 1)
            
            sx_term = self.sx
            if left_dims > 1:
                left = sp.eye(left_dims, dtype=complex)
                sx_term = sp.kron(left, sx_term)
            if right_dims > 1:
                right = sp.eye(right_dims, dtype=complex)
                sx_term = sp.kron(sx_term, right)
                
            H -= self.h * sx_term
            
        return H
    
    def sample_coherent_states_optimized(self):
        """Muestreo optimizado de estados coherentes"""
        print("Muestreando estados coherentes...")
        
        # Muestreo uniforme en la esfera S²
        u = np.random.uniform(-1, 1, self.S)
        phi = np.random.uniform(0, 2*np.pi, self.S)
        theta = np.arccos(u)
        
        # Coordenadas estereográficas
        z_samples = np.tan(theta/2) * np.exp(1j * phi)
        
        return z_samples
    
    def build_coherent_state_vector(self, z):
        """Estado coherente |z⟩ = (1/√(1+|z|²)) [1, z]^T"""
        norm = 1.0 / np.sqrt(1.0 + np.abs(z)**2)
        return np.array([1.0, z], dtype=complex) * norm
    
    def build_product_coherent_state(self, z):
        """Construye estado producto coherente |Z⟩ = ⊗_{i=1}^N |z⟩ (homogéneo)"""
        # Para estado homogéneo, todos los sitios tienen el mismo z
        single_site = self.build_coherent_state_vector(z)
        state = single_site
        for _ in range(self.N - 1):
            state = np.kron(state, single_site)
        return state
    
    def compute_gram_matrix_direct(self, z_samples):
        """Calcula matriz Gram directamente usando la estructura geométrica"""
        print("Calculando matriz Gram...")
        
        G = np.zeros((self.S, self.S), dtype=complex)
        factors = 1.0 / np.sqrt(1.0 + np.abs(z_samples)**2)
        
        for i in tqdm(range(self.S), desc="Gram matrix"):
            for j in range(self.S):
                # Producto interno ⟨z_i|z_j⟩ = (1 + z̄_i z_j) / √((1+|z_i|²)(1+|z_j|²))
                overlap = (1 + np.conj(z_samples[i]) * z_samples[j]) * factors[i] * factors[j]
                # Para producto de N espines: ⟨Z_i|Z_j⟩ = [⟨z_i|z_j⟩]^N
                G[i, j] = overlap ** self.N
                
        return G
    
    def build_projection_from_gram_optimized(self, G):
        """Construye proyección usando descomposición espectral"""
        print("Construyendo proyección...")
        
        # Diagonalización de G
        eigenvalues, eigenvectors = eigh(G)
        
        # Orden descendente
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]
        
        # Selección de dimensión truncada
        M_eff = min(self.M_trunc, np.sum(eigenvalues > 1e-12))
        
        print(f"Dimensión efectiva del subespacio: {M_eff}")
        return eigenvectors[:, :M_eff], eigenvalues[:M_eff]
    
    def build_data_matrix_optimized(self, z_samples):
        """Construye matriz de datos para obtener la base completa"""
        print("Construyendo matriz de datos...")
        
        Data = np.zeros((self.dim, self.S), dtype=complex)
        
        for i, z in enumerate(tqdm(z_samples, desc="Estados coherentes")):
            Data[:, i] = self.build_product_coherent_state(z)
            
        return Data
    
    def compute_effective_hamiltonian(self, H_sparse, P):
        """Calcula Hamiltoniano efectivo en el subespacio proyectado"""
        print("Calculando Hamiltoniano efectivo...")
        
        # H_eff = P^† H P
        H_eff = P.conj().T @ (H_sparse @ P)
        
        return H_eff
    
    def time_evolution_subspace(self, H_eff, initial_state_proj, t_points):
        """Evolución temporal en el subespacio reducido"""
        print("Calculando evolución temporal...")
        
        # Diagonalización del Hamiltoniano efectivo
        eigvals, eigvecs = eigh(H_eff)
        
        states_subspace = np.zeros((len(t_points), H_eff.shape[0]), dtype=complex)
        
        for i, t in enumerate(tqdm(t_points, desc="Evolución")):
            # exp(-i H_eff t) usando diagonalización
            phase_factors = np.exp(-1j * eigvals * t)
            evolution_op = eigvecs @ np.diag(phase_factors) @ eigvecs.conj().T
            states_subspace[i] = evolution_op @ initial_state_proj
            
        return states_subspace
    
    def compute_magnetization_observable(self, N):
        """Construye observable de magnetización total"""
        # ⟨σ_z^total⟩ = Σ_i ⟨σ_z^i⟩
        magnetizations = []
        for i in range(N):
            # Construir operador σ_z en el sitio i
            ops = [self.id2] * N
            ops[i] = self.sz
            op = ops[0]
            for j in range(1, N):
                op = np.kron(op, ops[j])
            magnetizations.append(op)
        
        def observable(state):
            total = 0
            for op in magnetizations:
                total += np.real(state.conj() @ (op @ state))
            return total / N
        
        return observable
    
    def run_analysis(self, t_max=5.0, n_t=100):
        """Ejecuta análisis completo"""
        print("=== ANÁLISIS DE PROYECCIÓN COHERENTE ESCALABLE ===")
        print(f"Parámetros: N={self.N}, S={self.S}, M_trunc={self.M_trunc}")
        
        # 1. Construir Hamiltonianos
        H_sparse = self.build_ising_hamiltonian_sparse_optimized()
        
        # 2. Muestrear estados coherentes
        z_samples = self.sample_coherent_states_optimized()
        
        # 3. Construir matriz Gram y proyección
        G = self.compute_gram_matrix_direct(z_samples)
        V, singular_values = self.build_projection_from_gram_optimized(G)
        
        # 4. Construir matriz de datos y base completa
        Data = self.build_data_matrix_optimized(z_samples)
        P = Data @ V  # Base en el espacio completo
        
        # 5. Hamiltoniano efectivo
        H_eff = self.compute_effective_hamiltonian(H_sparse, P)
        
        # 6. Estado inicial |00...0⟩
        psi0 = np.zeros(self.dim, dtype=complex)
        psi0[0] = 1.0
        psi0_proj = P.conj().T @ psi0
        
        # 7. Evolución temporal
        t_points = np.linspace(0, t_max, n_t)
        states_subspace = self.time_evolution_subspace(H_eff, psi0_proj, t_points)
        
        # 8. Observables
        magnetization_obs = self.compute_magnetization_observable(self.N)
        magnetization = np.zeros(n_t)
        
        for i in range(n_t):
            state_full = P @ states_subspace[i]
            magnetization[i] = magnetization_obs(state_full)
        
        # 9. Visualización corregida
        self.plot_results(t_points, magnetization, singular_values)
        
        return {
            't_points': t_points,
            'magnetization': magnetization,
            'singular_values': singular_values,
            'H_eff': H_eff,
            'P': P,
            'states_subspace': states_subspace
        }
    
    def plot_results(self, t_points, magnetization, singular_values):
        """Visualización corregida de resultados"""
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        # Verificar dimensiones
        print(f"Dimensiones - t_points: {len(t_points)}, magnetization: {len(magnetization)}")
        print(f"Dimensiones - singular_values: {len(singular_values)}")
        
        # Evolución de magnetización
        axes[0].plot(t_points, magnetization, 'b-', linewidth=2)
        axes[0].set_xlabel('Tiempo')
        axes[0].set_ylabel('Magnetización ⟨σ_z⟩')
        axes[0].set_title('Evolución Temporal')
        axes[0].grid(True, alpha=0.3)
        
        # Valores singulares
        x_singular = np.arange(len(singular_values))
        axes[1].semilogy(x_singular, singular_values, 'ro-', linewidth=2, markersize=4)
        axes[1].set_xlabel('Modo')
        axes[1].set_ylabel('Valor Singular')
        axes[1].set_title('Espectro de la Matriz Gram')
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

# Ejemplo de uso CORREGIDO
if __name__ == "__main__":
    # Sistema más pequeño para prueba rápida
    analyzer = ScalableCoherentProjection(N=6, S=100, M_trunc=30)
    results = analyzer.run_analysis(t_max=3.0, n_t=50)
```

# 2. Sobre el Flujo de Ricci 
La fidelidad $\mathcal{F}$ mide qué tan bien el estado inicial $\psi_0$ está contenido (o capturado) dentro del subespacio de baja dimensión $M$ que se construye en cada paso del flujo, pudiendo implementarse de la siguiente manera,

La función utilizada calcula 
$$
\mathcal{F} = \left| \langle \psi_0 | P P^H | \psi_0 \rangle \right|^2
$$

Donde
- $\psi_0 = |00...0\rangle$ es el estado inicial (todo \textit{up} en el código).
- $P$ es la matriz de proyección (de dimensión $D \times M$), cuyas columnas forman la base ortonormal del subespacio de dimensión $M$.
- $P P^H$ es el operador proyector sobre el subespacio $M$.

Si $\mathcal{F} \approx 1$, el subespacio $M$ es excelente para representar $\psi_0$. En cambio, si $\mathcal{F} \approx 0$, el subespacio $M$ es una mala elección de base para $\psi_0$.

---

## 2.1. Parámetro $\beta$ como escala geométrica.

El Flujo de Ricci se simula al variar $\beta$, que actúa como una \textbf{escala de resolución geométrica} o un tiempo de flujo ($\tau$) en el contexto de la teoría de campos.

El proyector $P(\beta)$ se construye utilizando el \textbf{Kernel Ponderado}, donde los pesos $w_k$ son proporcionales a

$$
w_k(\beta) \propto \langle Z_k | e^{-\beta H} | Z_k \rangle
$$

- \textbf{$\beta \approx 0$ (Alta Temperatura / Escala Larga):} Cuando $\beta$ es muy pequeño (alta temperatura), $e^{-\beta H} \approx \mathbb{I}$ (la identidad). Los pesos $w_k$ son casi uniformes. El subespacio $P(0)$ se define por la métrica geométrica "desnuda" sin tener en cuenta la dinámica del Hamiltoniano $H$.
- \textbf{$\beta$ grande (Baja Temperatura / Escala Corta):} $e^{-\beta H}$ proyecta fuertemente sobre los autoestados de baja energía de $H$. El proyector $P(\beta)$ da más peso a los estados coherentes $|Z_k\rangle$ que están cerca del \textit{manifold} de baja energía de $H$. Es la métrica geométrica "vestida", adaptada a la dinámica del sistema.

---

## 2.2. Interpretación de la curva $\mathcal{F}(\beta)$


La gráfica de $\mathcal{F}$ vs. $\beta$ constituye una herramienta diagnóstica fundamental, pues revela la eficiencia y la convergencia del flujo de Ricci hacia el subespacio de baja energía invariante definido por el Hamiltoniano $H$. La función de fidelidad $\mathcal{F}: \beta \mapsto [0, 1]$ se analiza rigurosamente mediante la división en tres regímenes distintos, análogos a las escalas en la física térmica o el flujo del grupo de renormalización.

### 2.2.1. $\beta \approx 0$: Límite Geométrico y de Altas Energías

En el límite de $\beta \to 0$, la temperatura efectiva $T \sim 1/\beta$ es infinita, siendo un límite geométrico y de altas energías. El factor de ponderación del kernel, $w_k(\beta) \propto \langle Z_k | e^{-\beta H} | Z_k \rangle$, se aproxima a la identidad, $w_k(\beta) \approx \langle Z_k | \mathbb{I} | Z_k \rangle = 1$, asumiendo la normalización del \textit{manifold}. Consecuentemente, el proyector $P(0)$ se construye exclusivamente a partir de la estructura geométrica del \textit{manifold} de estados coherentes, desacoplado de las propiedades espectrales de $H$. La fidelidad $\mathcal{F}(0)$ es típicamente baja o moderada, ya que el subespacio $M$ resultante es una representación puramente geométrica, no optimizada para capturar las componentes de baja energía del estado inicial $\psi_0$.

### 2.2.2. $\beta$ Intermedio: Régimen de Flujo y Optimización Métrica


A medida que $\beta$ crece desde cero, el operador exponencial $e^{-\beta H}$ comienza a diferenciar las ponderaciones $w_k(\beta)$, siendo un régimen de energías de flujo y optimización métrica. Los estados $|Z_k\rangle$ que poseen una mayor superposición con los autoestados de menor energía de $H$ (aquellos con autovalores $\lambda_n$ pequeños, haciendo $e^{-\beta \lambda_n}$ grande) experimentan una amplificación exponencial en sus pesos. Físicamente, esto indica que la estructura métrica del subespacio cuántico está siendo activamente \textit{vestida} o \textit{adaptada} por la dinámica hamiltoniana. El rápido aumento de $\mathcal{F}(\beta)$ en esta región demuestra la eficiencia del ponderado de tipo Flujo de Ricci para rotar a la sub-variedad diferencial $M$ hacia el subespacio de baja energía dinámicamente relevante, logrando una captura significativamente mejor del estado inicial $\psi_0$.

### 2.2.3. $\beta$ Grande: Límite del Estado Fundamental y de Baja Energía


En el límite de $\beta \to \infty$ (temperatura cero), $e^{-\beta H}$ se comporta asintóticamente como un proyector perfecto sobre el estado fundamental $\psi_g$ o el subespacio degenerado de baja energía. El kernel ponderado filtra rigurosamente las contribuciones de alta energía, lo que obliga a las columnas de $P(\beta)$ a abarcar el subespacio de menor energía de $H$. La estabilización de la fidelidad, $\mathcal{F}(\beta) \to \mathcal{F}_{\text{max}} \le 1$, señala la convergencia exitosa del sub-manifold $M$ al subespacio invariante verdadero de $H$. Si el estado inicial $\psi_0$ tiene una fuerte superposición con este \textit{manifold} de baja energía, $\mathcal{F}_{\text{max}}$ tenderá a la unidad, confirmando que el sub-manifold $M$ proporciona una representación óptima y dinámicamente adaptada de la física a largo plazo del sistema.

En resumen, la curva $\mathcal{F}(\beta)$ describe, ¿qué tan rápido y qué tan bien la geometría del subespacio, guiada por el Hamiltoniano $H$ (a través del Kernel Ponderado), se auto-adapta para capturar el estado $\psi_0$? 

Un aumento pronunciado significa que el flujo de $\beta$ refina eficientemente el subespacio hacia la región de interés.

```python
import numpy as np
import scipy.sparse as sp
from scipy.linalg import eigh, expm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
from matplotlib import cm
from tqdm import tqdm
import time

# =============================================================================
# CONFIGURACIÓN MEJORADA DEL SISTEMA
# =============================================================================

class RicciFlowConfig:
    """Configuración centralizada para el flujo de Ricci"""
    def __init__(self):
        # Parámetros físicos
        self.N = 8
        self.J = 1.0
        self.h = 0.8
        self.S = 500
        self.M_trunc = 50
        self.BETA_MAX = 5.0
        
        # Parámetros de visualización
        self.plot_3d_flow = True
        self.plot_heatmaps = True
        self.plot_fidelity_landscape = True
        self.plot_comparison = True
        self.dpi = 150
        
        # Parámetros de optimización
        self.use_progress_bars = True
        self.parallel_processing = False

# =============================================================================
# OPERADORES Y FUNCIONES MEJORADAS
# =============================================================================

class QuantumGeometry:
    """Geometría cuántica y operadores mejorados"""
    
    # Operadores de Pauli
    sx = np.array([[0.0,1.0],[1.0,0.0]], dtype=complex)
    sy = np.array([[0.0,-1j],[1j,0.0]], dtype=complex)
    sz = np.array([[1.0,0.0],[0.0,-1.0]], dtype=complex)
    id2 = np.eye(2, dtype=complex)
    
    @staticmethod
    def kron_list_sparse_optimized(mats):
        """Kronecker product optimizado para sparse matrices"""
        if len(mats) == 0:
            return sp.eye(1, format='csr', dtype=complex)
        
        result = mats[0]
        for mat in mats[1:]:
            result = sp.kron(result, mat, format='csr')
        return result
    
    @staticmethod
    def ising_hamiltonian_enhanced(N, J=1.0, h=0.5, periodic=False):
        """Hamiltoniano de Ising mejorado con más información"""
        dim = 2**N
        H = sp.csr_matrix((dim, dim), dtype=complex)
        
        # Interacción ZZ
        for i in range(N-1 + int(periodic)):
            j = (i+1) % N
            ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
            ops[i] = sp.csr_matrix(QuantumGeometry.sz)
            ops[j] = sp.csr_matrix(QuantumGeometry.sz)
            H -= J * QuantumGeometry.kron_list_sparse_optimized(ops)
        
        # Campo transversal
        for i in range(N):
            ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
            ops[i] = sp.csr_matrix(QuantumGeometry.sx)
            H -= h * QuantumGeometry.kron_list_sparse_optimized(ops)
        
        return H
    
    @staticmethod
    def build_observable_operators(N, observable='sz'):
        """Construye operadores observables para todo el sistema"""
        operators = []
        op_map = {'sz': QuantumGeometry.sz, 'sx': QuantumGeometry.sx, 'sy': QuantumGeometry.sy}
        base_op = op_map.get(observable, QuantumGeometry.sz)
        
        for i in range(N):
            ops = [QuantumGeometry.id2] * N
            ops[i] = base_op
            operator = QuantumGeometry.kron_list_dense(ops)
            operators.append(operator)
        
        return operators
    
    @staticmethod
    def kron_list_dense(mats):
        """Kronecker product para matrices densas"""
        result = mats[0]
        for mat in mats[1:]:
            result = np.kron(result, mat)
        return result

# =============================================================================
# MUESTREO Y ESTADOS COHERENTES MEJORADOS
# =============================================================================

class CoherentStateSampler:
    """Muestreo mejorado de estados coherentes"""
    
    @staticmethod
    def spin_coherent_state_enhanced(z):
        """Estado coherente mejorado con validación"""
        if np.isnan(z) or np.isinf(z):
            return np.array([1.0, 0.0], dtype=complex)
        
        norm_sq = 1.0 + np.abs(z)**2
        if norm_sq < 1e-12:
            return np.array([1.0, 0.0], dtype=complex)
            
        norm = 1.0 / np.sqrt(norm_sq)
        return np.array([1.0, z], dtype=complex) * norm
    
    @staticmethod
    def sample_cp1_stratified(count, method='haar'):
        """
        Muestreo estratificado de CP¹ con diferentes métodos
        """
        if method == 'haar':
            # Muestreo Haar uniforme tradicional
            u = np.random.uniform(-1.0, 1.0, size=count)
            phi = np.random.uniform(0.0, 2*np.pi, size=count)
            theta = np.arccos(u)
            
        elif method == 'fibonacci':
            # Muestreo Fibonacci para distribución más uniforme
            indices = np.arange(0, count, dtype=float)
            phi = 2 * np.pi * indices * (1.0 / ((1.0 + np.sqrt(5)) / 2.0))
            theta = np.arccos(1.0 - 2.0 * (indices + 0.5) / count)
            
        elif method == 'grid':
            # Malla estructurada
            n_theta = int(np.sqrt(count))
            n_phi = int(count / n_theta)
            theta = np.arccos(1.0 - 2.0 * (np.arange(n_theta) + 0.5) / n_theta)
            phi = 2 * np.pi * np.arange(n_phi) / n_phi
            theta, phi = np.meshgrid(theta, phi)
            theta, phi = theta.flatten(), phi.flatten()
            
        z = np.tan(theta/2.0) * np.exp(1j*phi)
        return z[:count], theta[:count], phi[:count]
    
    @staticmethod
    def build_heterogeneous_coherent_vectors_optimized(N, S, sampling_method='haar'):
        """
        Construcción optimizada de estados producto heterogéneos
        """
        dim = 2**N
        Data = np.zeros((dim, S), dtype=complex)
        
        # Muestreo más inteligente
        z_samples_all, theta_all, phi_all = CoherentStateSampler.sample_cp1_stratified(S * N, sampling_method)
        
        for k in tqdm(range(S), desc="Construyendo estados coherentes"):
            z_sample_k = z_samples_all[k*N : (k+1)*N]
            vec = CoherentStateSampler.spin_coherent_state_enhanced(z_sample_k[0])
            
            for i in range(1, N):
                single_site = CoherentStateSampler.spin_coherent_state_enhanced(z_sample_k[i])
                vec = np.kron(vec, single_site)
                
            Data[:, k] = vec / np.linalg.norm(vec)  # Normalización adicional
            
        return Data, theta_all, phi_all

# =============================================================================
# KERNEL TÉRMICO Y PESOS MEJORADOS
# =============================================================================

class ThermalKernel:
    """Kernel térmico mejorado para el flujo de Ricci"""
    
    @staticmethod
    def get_thermal_weights_optimized(H_sparse, Data, beta, method='exact'):
        """
        Pesos térmicos optimizados con diferentes métodos de aproximación
        """
        if beta < 1e-8:
            return np.ones(Data.shape[1], dtype=float) / Data.shape[1]
        
        if method == 'exact':
            # Método exacto (para sistemas pequeños)
            H_dense = H_sparse.todense()
            P_op = expm(-beta * H_dense)
            
        elif method == 'lanczos':
            # Aproximación con Lanczos (para sistemas grandes)
            return ThermalKernel._lanczos_thermal_weights(H_sparse, Data, beta)
        
        weights = np.zeros(Data.shape[1], dtype=float)
        
        for k in range(Data.shape[1]):
            Zk = Data[:, k]
            weights[k] = np.real(Zk.conj() @ (P_op @ Zk))
            
        # Suavizado y normalización
        weights = np.maximum(weights, 1e-12)  # Evitar ceros
        weights /= np.sum(weights)
        
        return weights
    
    @staticmethod
    def _lanczos_thermal_weights(H_sparse, Data, beta, k=10):
        """Aproximación de pesos usando Lanczos (placeholder para sistemas grandes)"""
        # Implementación simplificada - usar método exacto por ahora
        return ThermalKernel.get_thermal_weights_optimized(H_sparse, Data, beta, method='exact')
    
    @staticmethod
    def compute_ricci_curvature_approximation(weights, beta):
        """
        Aproximación de la curvatura de Ricci a partir de la distribución de pesos
        """
        entropy = -np.sum(weights * np.log(weights + 1e-12))
        ricci_approx = (np.log(len(weights)) - entropy) / beta if beta > 1e-8 else 0.0
        return ricci_approx, entropy

# =============================================================================
# PROYECCIÓN Y ANÁLISIS GEOMÉTRICO MEJORADO
# =============================================================================

class GeometricProjection:
    """Análisis geométrico mejorado de la proyección"""
    
    @staticmethod
    def build_weighted_projection_enhanced(Data, M, weights, regularization=1e-8):
        """
        Proyección ponderada mejorada con regularización
        """
        sqrt_weights = np.sqrt(weights)
        D_tilde = Data * sqrt_weights[np.newaxis, :]
        
        # Matriz Gram con regularización
        G_tilde = D_tilde.conj().T @ D_tilde 
        G_tilde += regularization * np.eye(G_tilde.shape[0])
        
        # Descomposición espectral
        evals, evecs = eigh(G_tilde)
        idx = np.argsort(evals)[::-1]
        evals, evecs = evals[idx], evecs[:, idx]
        
        # Selección adaptativa de dimensión
        M_eff = GeometricProjection._adaptive_dimension_selection(evals, M)
        
        # Construcción de la base proyectada
        P = np.zeros((Data.shape[0], M_eff), dtype=complex)
        for m in range(M_eff):
            v = evecs[:, m]
            lam = evals[m]
            P[:, m] = (D_tilde @ v) / np.sqrt(lam)
            
        return P, evals, M_eff
    
    @staticmethod
    def _adaptive_dimension_selection(eigenvalues, M_max, threshold_ratio=0.01):
        """Selección adaptativa de la dimensión basada en espectro"""
        total_variance = np.sum(eigenvalues)
        cumulative_variance = np.cumsum(eigenvalues) / total_variance
        
        # Encontrar donde la varianza acumulada supera el umbral
        M_adaptive = np.argmax(cumulative_variance > (1 - threshold_ratio)) + 1
        return min(M_adaptive, M_max, len(eigenvalues))
    
    @staticmethod
    def analyze_projection_quality(P, H, psi0, Data):
        """Análisis completo de la calidad de la proyección"""
        # Fidelidad inicial
        fid_initial = GeometricProjection.calculate_initial_fidelity(P, psi0)
        
        # Error de proyección del Hamiltoniano
        H_proj = P.conj().T @ H @ P
        H_reconstructed = P @ H_proj @ P.conj().T
        hamiltonian_error = np.linalg.norm(H.todense() - H_reconstructed, 'fro')
        
        # Calidad geométrica de la base
        gram_matrix = P.conj().T @ P
        orthogonality_error = np.linalg.norm(gram_matrix - np.eye(gram_matrix.shape[0]), 'fro')
        
        return {
            'fidelity': fid_initial,
            'hamiltonian_error': hamiltonian_error,
            'orthogonality_error': orthogonality_error,
            'effective_dimension': P.shape[1]
        }
    
    @staticmethod
    def calculate_initial_fidelity(P_matrix, psi0):
        """Cálculo robusto de fidelidad"""
        psi0_proj = P_matrix.conj().T @ psi0
        psi0_reconstructed = P_matrix @ psi0_proj
        overlap = np.vdot(psi0_reconstructed, psi0)
        return np.abs(overlap)**2

# =============================================================================
# VISUALIZACIÓN AVANZADA DEL FLUJO DE RICCI
# =============================================================================

class RicciFlowVisualizer:
    """Visualizaciones avanzadas para el flujo de Ricci"""
    
    @staticmethod
    def plot_ricci_flow_3d(beta_list, fidelities, curvatures, entropies):
        """Visualización 3D del flujo de Ricci"""
        fig = plt.figure(figsize=(15, 5))
        
        # Subplot 1: Fidelidad vs Beta
        ax1 = fig.add_subplot(131)
        ax1.plot(beta_list, fidelities, 'o-', linewidth=2.5, color='#8b5cf6', markersize=6)
        ax1.set_xlabel(r'Parámetro de Escala $\beta$', fontsize=12)
        ax1.set_ylabel('Fidelidad $\mathcal{F}$', fontsize=12)
        ax1.set_title('Evolución de Fidelidad', fontsize=14)
        ax1.grid(True, alpha=0.3)
        ax1.set_ylim(0, 1.1)
        
        # Subplot 2: Curvatura de Ricci aproximada
        ax2 = fig.add_subplot(132)
        ax2.plot(beta_list, curvatures, 's-', linewidth=2.5, color='#ef4444', markersize=6)
        ax2.set_xlabel(r'Parámetro de Escala $\beta$', fontsize=12)
        ax2.set_ylabel('Curvatura de Ricci Aprox.', fontsize=12)
        ax2.set_title('Flujo de Curvatura', fontsize=14)
        ax2.grid(True, alpha=0.3)
        
        # Subplot 3: Entropía térmica
        ax3 = fig.add_subplot(133)
        ax3.plot(beta_list, entropies, '^-', linewidth=2.5, color='#10b981', markersize=6)
        ax3.set_xlabel(r'Parámetro de Escala $\beta$', fontsize=12)
        ax3.set_ylabel('Entropía Térmica', fontsize=12)
        ax3.set_title('Evolución de Entropía', fontsize=14)
        ax3.grid(True, alpha=0.3)
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_fidelity_landscape(beta_list, fidelities, qualities):
        """Landscape de fidelidad con métricas de calidad"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
        
        # Landscape principal
        im = ax1.scatter(beta_list, fidelities, c=qualities, 
                        cmap='viridis', s=100, alpha=0.7)
        ax1.set_xlabel(r'$\beta$')
        ax1.set_ylabel('Fidelidad')
        ax1.set_title('Landscape de Fidelidad')
        plt.colorbar(im, ax=ax1, label='Calidad de Proyección')
        
        # Análisis de derivada
        derivative = np.gradient(fidelities, beta_list)
        ax2.plot(beta_list, derivative, 'r-', linewidth=2)
        ax2.axhline(0, color='k', linestyle='--', alpha=0.5)
        ax2.set_xlabel(r'$\beta$')
        ax2.set_ylabel('Derivada de Fidelidad')
        ax2.set_title('Análisis de Sensibilidad')
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        return fig
    
    @staticmethod
    def plot_projection_heatmaps(P_best, H_eff, beta_optimal):
        """Mapas de calor de la proyección óptima"""
        fig, axes = plt.subplots(1, 3, figsize=(15, 5))
        
        # Matriz de proyección
        im1 = axes[0].imshow(np.abs(P_best), aspect='auto', cmap='hot')
        axes[0].set_title(f'Matriz de Proyección (β={beta_optimal:.2f})')
        axes[0].set_xlabel('Dimensión Reducida')
        axes[0].set_ylabel('Dimensión Completa')
        plt.colorbar(im1, ax=axes[0])
        
        # Hamiltoniano efectivo (parte real)
        im2 = axes[1].imshow(H_eff.real, aspect='auto', cmap='RdBu_r')
        axes[1].set_title('Hamiltoniano Efectivo (Real)')
        axes[1].set_xlabel('Base Reducida')
        axes[1].set_ylabel('Base Reducida')
        plt.colorbar(im2, ax=axes[1])
        
        # Hamiltoniano efectivo (parte imaginaria)
        im3 = axes[2].imshow(H_eff.imag, aspect='auto', cmap='RdBu_r')
        axes[2].set_title('Hamiltoniano Efectivo (Imag)')
        axes[2].set_xlabel('Base Reducida')
        axes[2].set_ylabel('Base Reducida')
        plt.colorbar(im3, ax=axes[2])
        
        plt.tight_layout()
        return fig

# =============================================================================
# FLUJO DE RICCI MEJORADO
# =============================================================================

class EnhancedRicciFlow:
    """Implementación mejorada del flujo de Ricci"""
    
    def __init__(self, config):
        self.config = config
        self.H = None
        self.Data = None
        self.results = {}
        
    def initialize_system(self):
        """Inicialización del sistema cuántico"""
        print("=== INICIALIZANDO SISTEMA CUÁNTICO ===")
        
        # Construir Hamiltoniano
        self.H = QuantumGeometry.ising_hamiltonian_enhanced(
            self.config.N, self.config.J, self.config.h)
        
        dim = 2**self.config.N
        print(f"Sistema: N={self.config.N}, Dim={dim}")
        
        # Muestreo de estados coherentes
        self.Data, theta, phi = CoherentStateSampler.build_heterogeneous_coherent_vectors_optimized(
            self.config.N, self.config.S, 'haar')
        
        print(f"Muestreo: {self.config.S} estados coherentes generados")
        return self.H, self.Data
    
    def analyze_ricci_flow_enhanced(self, beta_list):
        """Análisis mejorado del flujo de Ricci"""
        print("\n=== ANALIZANDO FLUJO DE RICCI MEJORADO ===")
        
        dim = self.H.shape[0]
        psi0 = np.zeros((dim,), dtype=complex)
        psi0[0] = 1.0  # Estado inicial |00...0⟩
        
        # Métricas a trackear
        fidelities = []
        curvatures = []
        entropies = []
        qualities = []
        projections = []
        
        beta_iter = tqdm(beta_list, desc="Flujo de Ricci") if self.config.use_progress_bars else beta_list
        
        for beta in beta_iter:
            start_time = time.time()
            
            # 1. Pesos térmicos
            weights = ThermalKernel.get_thermal_weights_optimized(self.H, self.Data, beta)
            
            # 2. Proyección ponderada
            P_beta, evals, M_eff = GeometricProjection.build_weighted_projection_enhanced(
                self.Data, self.config.M_trunc, weights)
            
            # 3. Análisis de calidad
            quality_metrics = GeometricProjection.analyze_projection_quality(P_beta, self.H, psi0, self.Data)
            
            # 4. Curvatura y entropía
            ricci_curv, entropy = ThermalKernel.compute_ricci_curvature_approximation(weights, beta)
            
            # Almacenar resultados
            fidelities.append(quality_metrics['fidelity'])
            curvatures.append(ricci_curv)
            entropies.append(entropy)
            qualities.append(quality_metrics['hamiltonian_error'])
            projections.append(P_beta)
            
            elapsed = time.time() - start_time
            if self.config.use_progress_bars:
                beta_iter.set_postfix({
                    'β': f'{beta:.2f}',
                    'Fid': f'{quality_metrics["fidelity"]:.4f}',
                    'M_eff': M_eff
                })
            else:
                print(f"β={beta:.2f}: Fid={quality_metrics['fidelity']:.4f}, M_eff={M_eff}, t={elapsed:.2f}s")
        
        # Encontrar proyección óptima
        best_idx = np.argmax(fidelities)
        best_beta = beta_list[best_idx]
        best_P = projections[best_idx]
        best_H_eff = best_P.conj().T @ self.H @ best_P
        
        self.results = {
            'beta_list': beta_list,
            'fidelities': np.array(fidelities),
            'curvatures': np.array(curvatures),
            'entropies': np.array(entropies),
            'qualities': np.array(qualities),
            'best_beta': best_beta,
            'best_fidelity': fidelities[best_idx],
            'best_P': best_P,
            'best_H_eff': best_H_eff,
            'projections': projections
        }
        
        return self.results
    
    def generate_comprehensive_visualizations(self):
        """Genera todas las visualizaciones"""
        if not self.results:
            print("No hay resultados para visualizar")
            return
        
        print("\n=== GENERANDO VISUALIZACIONES COMPLETAS ===")
        
        # 1. Flujo de Ricci 3D
        if self.config.plot_3d_flow:
            fig1 = RicciFlowVisualizer.plot_ricci_flow_3d(
                self.results['beta_list'],
                self.results['fidelities'],
                self.results['curvatures'],
                self.results['entropies']
            )
            plt.savefig('ricci_flow_3d.png', dpi=self.config.dpi, bbox_inches='tight')
        
        # 2. Landscape de fidelidad
        if self.config.plot_fidelity_landscape:
            fig2 = RicciFlowVisualizer.plot_fidelity_landscape(
                self.results['beta_list'],
                self.results['fidelities'],
                self.results['qualities']
            )
            plt.savefig('fidelity_landscape.png', dpi=self.config.dpi, bbox_inches='tight')
        
        # 3. Mapas de calor de la proyección óptima
        if self.config.plot_heatmaps:
            fig3 = RicciFlowVisualizer.plot_projection_heatmaps(
                self.results['best_P'],
                self.results['best_H_eff'],
                self.results['best_beta']
            )
            plt.savefig('optimal_projection_heatmaps.png', dpi=self.config.dpi, bbox_inches='tight')
        
        plt.show()
        
        # Resumen numérico
        self._print_summary()
    
    def _print_summary(self):
        """Imprime resumen numérico de los resultados"""
        print(f"\n=== RESUMEN DEL FLUJO DE RICCI ===")
        print(f"Mejor β: {self.results['best_beta']:.4f}")
        print(f"Mejor fidelidad: {self.results['best_fidelity']:.6f}")
        print(f"Dimensión efectiva: {self.results['best_P'].shape[1]}")
        print(f"Rango de fidelidad: {np.min(self.results['fidelities']):.4f} - {np.max(self.results['fidelities']):.4f}")
        print(f"Curvatura máxima: {np.max(self.results['curvatures']):.4f}")

# =============================================================================
# FUNCIÓN PRINCIPAL MEJORADA
# =============================================================================

def main_enhanced_ricci_flow():
    """Función principal mejorada"""
    
    # Configuración
    config = RicciFlowConfig()
    
    # Inicializar sistema
    ricci_flow = EnhancedRicciFlow(config)
    H, Data = ricci_flow.initialize_system()
    
    # Parámetros beta (más puntos para mejor resolución)
    beta_list = np.linspace(0.01, config.BETA_MAX, 15)
    
    # Ejecutar análisis
    results = ricci_flow.analyze_ricci_flow_enhanced(beta_list)
    
    # Generar visualizaciones
    ricci_flow.generate_comprehensive_visualizations()
    
    return results

if __name__ == "__main__":
    results = main_enhanced_ricci_flow()
```

```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.linalg import eigh
import matplotlib.pyplot as plt
from tqdm import tqdm
import time

class ScalableRicciFlow:
    """
    Implementación escalable y corregida del Flujo de Ricci
    """
    
    def __init__(self, N, J=1.0, h=0.8, S=200, M_trunc=30, beta_max=3.0):
        self.N = N
        self.J = J
        self.h = h
        self.S = S
        self.M_trunc = M_trunc
        self.beta_max = beta_max
        self.dim = 2**N
        
        # Operadores de Pauli
        self.sx = np.array([[0, 1], [1, 0]], dtype=complex)
        self.sz = np.array([[1, 0], [0, -1]], dtype=complex)
        self.id2 = np.eye(2, dtype=complex)
        
    def build_ising_hamiltonian_sparse(self):
        """Hamiltoniano de Ising optimizado"""
        H = sp.csr_matrix((self.dim, self.dim), dtype=complex)
        
        for i in range(self.N - 1):
            left_dims = 2**i
            middle_dims = 2**(self.N - i - 2)
            
            sz_sz = sp.kron(self.sz, self.sz)
            if left_dims > 1:
                left = sp.eye(left_dims, dtype=complex)
                sz_sz = sp.kron(left, sz_sz)
            if middle_dims > 1:
                right = sp.eye(middle_dims, dtype=complex)
                sz_sz = sp.kron(sz_sz, right)
                
            H -= self.J * sz_sz
        
        for i in range(self.N):
            left_dims = 2**i
            right_dims = 2**(self.N - i - 1)
            
            sx_term = self.sx
            if left_dims > 1:
                left = sp.eye(left_dims, dtype=complex)
                sx_term = sp.kron(left, sx_term)
            if right_dims > 1:
                right = sp.eye(right_dims, dtype=complex)
                sx_term = sp.kron(sx_term, right)
                
            H -= self.h * sx_term
            
        return H
    
    def sample_heterogeneous_coherent_states(self):
        """Muestreo de estados coherentes heterogéneos"""
        print("Muestreando estados coherentes heterogéneos...")
        
        z_samples_all = np.zeros((self.S, self.N), dtype=complex)
        
        for i in range(self.S):
            u = np.random.uniform(-1, 1, self.N)
            phi = np.random.uniform(0, 2*np.pi, self.N)
            theta = np.arccos(u)
            z_samples_all[i] = np.tan(theta/2) * np.exp(1j * phi)
            
        return z_samples_all
    
    def build_coherent_state_vector(self, z):
        """Estado coherente individual"""
        norm = 1.0 / np.sqrt(1.0 + np.abs(z)**2)
        return np.array([1.0, z], dtype=complex) * norm
    
    def build_product_state(self, z_array):
        """Construye estado producto para array de z's"""
        state = self.build_coherent_state_vector(z_array[0])
        for i in range(1, len(z_array)):
            single_site = self.build_coherent_state_vector(z_array[i])
            state = np.kron(state, single_site)
        return state
    
    def compute_thermal_weights_krylov(self, H_sparse, data_matrix, beta, k=10):
        """
        Calcula pesos térmicos usando método Krylov aproximado
        w_k ≈ ⟨Z_k|exp(-βH)|Z_k⟩
        """
        print(f"Calculando pesos térmicos (β={beta:.3f})...")
        
        weights = np.zeros(data_matrix.shape[1])
        
        for i in tqdm(range(data_matrix.shape[1]), desc="Pesos térmicos"):
            state = data_matrix[:, i]
            
            try:
                # Usar expm_multiply para cálculo eficiente
                from scipy.sparse.linalg import expm_multiply
                evolved_state = expm_multiply(-beta * H_sparse, state)
                weight = np.abs(np.vdot(state, evolved_state))
                weights[i] = max(weight, 1e-12)
            except Exception as e:
                # Fallback: aproximación de primer orden
                H_state = H_sparse @ state
                energy_expectation = np.real(np.vdot(state, H_state))
                weight = 1.0 - beta * energy_expectation
                weights[i] = max(weight, 1e-12)
        
        # Normalizar y suavizar
        weights = np.maximum(weights, 1e-12)
        weights /= np.sum(weights)
        
        return weights
    
    def compute_gram_matrix_heterogeneous(self, z_samples_all):
        """Matriz Gram para estados heterogéneos - CORREGIDA"""
        print("Calculando matriz Gram heterogénea...")
        
        G = np.zeros((self.S, self.S), dtype=complex)
        
        for i in tqdm(range(self.S), desc="Gram matrix"):
            for j in range(i, self.S):  # Aprovechar simetría
                overlap = 1.0
                for k in range(self.N):
                    z_i = z_samples_all[i, k]
                    z_j = z_samples_all[j, k]
                    factor_i = 1.0 / np.sqrt(1.0 + np.abs(z_i)**2)
                    factor_j = 1.0 / np.sqrt(1.0 + np.abs(z_j)**2)
                    site_overlap = (1 + np.conj(z_i) * z_j) * factor_i * factor_j
                    overlap *= site_overlap
                G[i, j] = overlap
                if i != j:
                    G[j, i] = np.conj(overlap)  # Hermiticidad
                
        return G
    
    def build_weighted_projection(self, G, weights, data_matrix):
        """Proyección ponderada corregida"""
        print("Construyendo proyección ponderada...")
        
        # Matriz Gram ponderada
        sqrt_weights = np.sqrt(weights)
        G_weighted = G * np.outer(sqrt_weights, sqrt_weights)
        
        # Diagonalización
        eigenvalues, eigenvectors = eigh(G_weighted)
        
        # Orden descendente
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]
        
        # Selección adaptativa
        M_eff = min(self.M_trunc, np.sum(eigenvalues > 1e-12))
        
        # Construir base en espacio completo
        P = np.zeros((self.dim, M_eff), dtype=complex)
        for m in range(M_eff):
            v = eigenvectors[:, m]
            lam = eigenvalues[m]
            if lam > 1e-12:
                P[:, m] = (data_matrix @ v) / np.sqrt(lam)
        
        print(f"Dimensión efectiva: {M_eff}")
        return P, eigenvalues[:M_eff]
    
    def compute_fidelity(self, P, psi0):
        """Calcula fidelidad F = |⟨ψ0|P P^†|ψ0⟩|²"""
        psi0_proj = P.conj().T @ psi0
        psi0_reconstructed = P @ psi0_proj
        fidelity = np.abs(np.vdot(psi0_reconstructed, psi0))**2
        return fidelity
    
    def analyze_ricci_flow(self, n_beta=10):
        """Análisis completo del flujo de Ricci - CORREGIDO"""
        print("=== ANÁLISIS DE FLUJO DE RICCI ESCALABLE ===")
        print(f"Parámetros: N={self.N}, S={self.S}, M_trunc={self.M_trunc}")
        
        # 1. Construir Hamiltonianos y datos
        H_sparse = self.build_ising_hamiltonian_sparse()
        z_samples_all = self.sample_heterogeneous_coherent_states()
        
        # Construir matriz de datos
        data_matrix = np.zeros((self.dim, self.S), dtype=complex)
        for i in range(self.S):
            data_matrix[:, i] = self.build_product_state(z_samples_all[i])
        
        # 2. Calcular matriz Gram base (una sola vez)
        G_base = self.compute_gram_matrix_heterogeneous(z_samples_all)
        
        # 3. Estado inicial
        psi0 = np.zeros(self.dim, dtype=complex)
        psi0[0] = 1.0
        
        # 4. Puntos beta para el flujo
        beta_list = np.linspace(0.01, self.beta_max, n_beta)
        
        # 5. Análisis para cada beta
        fidelities = []
        effective_dims = []
        all_projections = []
        
        print("Ejecutando flujo de Ricci...")
        for beta in tqdm(beta_list, desc="Flujo β"):
            try:
                # Pesos térmicos
                weights = self.compute_thermal_weights_krylov(H_sparse, data_matrix, beta)
                
                # Proyección ponderada
                P_beta, evals = self.build_weighted_projection(G_base, weights, data_matrix)
                
                # Fidelidad
                fidelity = self.compute_fidelity(P_beta, psi0)
                fidelities.append(fidelity)
                effective_dims.append(P_beta.shape[1])
                all_projections.append(P_beta)
                
            except Exception as e:
                print(f"Error en β={beta:.3f}: {e}")
                fidelities.append(0.0)
                effective_dims.append(0)
                all_projections.append(None)
        
        # 6. Encontrar beta óptimo
        valid_indices = [i for i, f in enumerate(fidelities) if f > 0]
        if valid_indices:
            best_idx = valid_indices[np.argmax(np.array(fidelities)[valid_indices])]
            best_beta = beta_list[best_idx]
            best_fidelity = fidelities[best_idx]
            best_P = all_projections[best_idx]
        else:
            best_beta = beta_list[0]
            best_fidelity = fidelities[0]
            best_P = all_projections[0]
        
        print(f"Mejor β: {best_beta:.3f}, Fidelidad: {best_fidelity:.6f}")
        
        # 7. Visualización corregida
        self.plot_ricci_flow_results(beta_list, fidelities, effective_dims)
        
        return {
            'beta_list': beta_list,
            'fidelities': np.array(fidelities),
            'effective_dims': np.array(effective_dims),
            'best_beta': best_beta,
            'best_fidelity': best_fidelity,
            'best_P': best_P,
            'all_projections': all_projections
        }
    
    def plot_ricci_flow_results(self, beta_list, fidelities, effective_dims):
        """Visualización corregida del flujo de Ricci"""
        # Verificar dimensiones
        print(f"Dimensiones - beta_list: {len(beta_list)}, fidelities: {len(fidelities)}")
        print(f"Dimensiones - effective_dims: {len(effective_dims)}")
        
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        # Fidelidad vs Beta - CORREGIDO
        axes[0].plot(beta_list, fidelities, 'o-', linewidth=2.5, color='#8b5cf6', markersize=6)
        axes[0].set_xlabel('Parámetro de Escala β')
        axes[0].set_ylabel('Fidelidad $\mathcal{F}$')
        axes[0].set_title('Flujo de Ricci: Fidelidad vs β')
        axes[0].grid(True, alpha=0.3)
        axes[0].set_ylim(0, 1.1)
        
        # Dimensión efectiva vs Beta - CORREGIDO
        axes[1].plot(beta_list, effective_dims, 's-', linewidth=2.5, color='#ef4444', markersize=6)
        axes[1].set_xlabel('Parámetro de Escala β')
        axes[1].set_ylabel('Dimensión Efectiva')
        axes[1].set_title('Flujo de Ricci: Dimensión vs β')
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

# Ejemplo de uso CORREGIDO
if __name__ == "__main__":
    # Sistema pequeño para prueba rápida
    ricci_flow = ScalableRicciFlow(N=6, S=100, M_trunc=25, beta_max=2.0)
    results = ricci_flow.analyze_ricci_flow(n_beta=8)
```
