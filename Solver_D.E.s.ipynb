## 1. Marco funcional y notación

Sea $N$ el número de espines. El espacio de Hilbert físico es

$$
H = (\mathbb{C}^2)^{\otimes N}, \quad \dim H = d := 2^N.
$$

Sea $H: H \to H$ el Hamiltoniano hermítico (por ejemplo de Ising).  El producto interno se denota $\langle \cdot, \cdot \rangle$ y la norma $\|\cdot\|$.

Objetivo: construir un subespacio de dimensión $M \ll d=2^N$, mediante una construcción geométrica basada en secciones holomorfas (estados coherentes), proyectar $H$ a ese subespacio y controlar el error espectral y dinámico, reduciendo las dimensiones y por ende, el procesamiento y memoria.

---

## 2. Geometría compleja de $\mathbb{C}P^1$ y fibrado tautológico

### 2.1 Espacio proyectivo complejo

El espacio proyectivo complejo $\mathbb{C}P^n$ se define como el conjunto de líneas complejas que pasan por el origen en $\mathbb{C}^{n+1}$:

$$
\mathbb{C}P^n = \{ [z_0 : z_1 : \cdots : z_n] \,|\, (z_0,\ldots,z_n)\neq 0 \},
$$

donde $[z_0:...:z_n]$ denota la clase de equivalencia bajo la relación $(z_0,\ldots,z_n) \sim \lambda(z_0,\ldots,z_n)$ para $\lambda \in \mathbb{C}^\times$.

Cada punto de $\mathbb{C}P^n$ corresponde a una línea compleja unidimensional en $\mathbb{C}^{n+1}$.

### 2.2 Fibrado tautológico

El **fibrado tautológico** $\mathcal{O}(-1)$ sobre $\mathbb{C}P^n$ se define mediante

$$
\mathcal{O}(-1) = \{ ([z], v) \in \mathbb{C}P^n \times \mathbb{C}^{n+1} \, | \, v \in [z] \}.
$$

Su fibra en el punto $[z]$ es precisamente la línea compleja generada por $z$, de forma tal que, su dual dual $\mathcal{O}(1)$ es el **fibrado hiperbólico**, que tiene un rol central en la representación de estados cuánticos y secciones holomorfas.

---

## 3. Geometría Kähler y métrica de Fubini–Study

El espacio $\mathbb{C}P^n$ posee una estructura Kähler natural, con métrica inducida por el potencial de Kähler

$$
K(z, \bar{z}) = \log(1 + |z|^2),
$$

en coordenadas inhomogéneas $z = (z_1, \ldots, z_n)$. La forma de Kähler asociada, siendo la métrica de Fubini–Study es

$$
\omega_{FS} = i \partial \bar{\partial} K = \frac{i \sum_{i,j} [(1+|z|^2)\delta_{ij} - \bar{z}_i z_j]\, dz_i \wedge d\bar{z}_j}{(1+|z|^2)^2}.
$$

La integral de $\omega_{FS}$ sobre una 2-esfera compleja (por ejemplo $\mathbb{C}P^1$ embebida) da el **número de Chern** de $\mathcal{O}(1)$

$$
\frac{1}{2\pi}\int_{\mathbb{C}P^1}\omega_{FS} = 1.
$$

Esto establece el vínculo topológico entre los fibrados complejos y la cuantización geométrica. Utilizando de forma general una 2-forma diferencial $\omega=i g_{i\bar{j}}dz^i \wedge d\bar{z}^j$, el flujo de Ricci del sistema evolucionaria en función de un parámetro $\tau$ como


$$
\partial_{\tau} \omega=i\partial\bar{\partial} \ln{\det{g_{i\bar{j}}(\tau)}}.
$$

siendo de importancia para la renormalización geométrica y el estudio de las singularidades del espacio de estados.

---

## 4. Estados coherentes y representación holomorfa

Identificamos la esfera $S^2$ con $\mathbb{C}P^1$ mediante la proyección estereográfica vía un homeomorfismo diferenciable. Para cada $z \in \mathbb{C}$, el estado coherente (sección local de $\mathcal{O}(1)$) es

$$
|z\rangle = \frac{1}{\sqrt{1+|z|^2}}
\begin{pmatrix}
1 \\ z
\end{pmatrix} \in \mathbb{C}^2.
$$

Propiedades

$$
\langle z | z \rangle = 1, \quad 
\int_{\mathbb{C}} |z\rangle\langle z| \, \frac{2}{\pi}\frac{d^2z}{(1+|z|^2)^2} = I_2.
$$

Para $N$ espines

$$
|Z\rangle = \bigotimes_{i=1}^N |z_i\rangle, \quad
\int |Z\rangle \langle Z|\, d\mu^{(N)}(Z) = I_H.
$$

Por ejemplo, para un único espin, la ecuación se verifica expandiendo $|z\rangle\langle z|$ en la base $\{\ket{0},\ket{1}\}$ y usando coordenadas polares mediante la integral 

$$
\int_{0}^{2\pi} e^{im\phi}d\phi =0, \  \, \forall m \not= 0.
$$

La integral radial reproduce las constantes necesarias.

---

## 5. Kernel de Bergman y espacio de secciones holomorfas

Consideremos el espacio de secciones holomorfas de $\mathcal{O}(k)$ sobre $\mathbb{C}P^1$:

$$
H_k = H^0(\mathbb{C}P^1, \mathcal{O}(k)) \simeq \mathrm{span}\{1, z, z^2, \ldots, z^k\}, \quad \dim H_k = k+1.
$$

El **kernel de Bergman** asociado es

$$
K_k(z, w) = \sum_{j=0}^{k} s_j(z) \overline{s_j(w)} \propto (1 + z\bar{w})^k.
$$

Con propiedad de reproducción 

$$
f(z) = \int_{\mathbb{C}P^1} K_k(z, w) f(w) \, d\mu(w), \quad \forall f \in H_k.
$$

Este kernel actúa como el análogo matemático del **kernel de Penrose** en la teoría twistorial.

---

## 6. Proyección discreta y método de Nyström

Tomemos un conjunto de $S$ puntos del fibrado, es decir, cada $Z_j \in \mathbb{C}^N$. Se define la matriz de datos

$$
D \in \mathbb{C}^{d \times S}, \quad D[:,j] = |Z_j\rangle, \quad G = D^\dagger D \in \mathbb{C}^{S\times S}, \quad G_{ij}=|Z_i\rangle\langle Z_j|
$$

El operador integral de una proyección continua $K$ actuando en $L^2$ se puede discretizar por puntos de muestreo para dar una aproximación finita-rango. En el entorno finito-dimeno, porque el espacio físico $\mathcal{H}$ ya tiene una dimensión finita $d$, la columna espacio de $D$ es precisamente el subespacio generado por las $S$ secciones muestreadas. La diagonalización de $G$ da la base ortonormal principal asociada al sistema. Concretamente, diagonalizando $G = V \Lambda V^\dagger$, definimos

$$
u_m = \frac{1}{\sqrt{\lambda_m}} D v_m, \quad P = \sum_{m=1}^M |u_m\rangle\langle u_m|.
$$

Entonces $P$ es una proyección ortogonal sobre un subespacio $S \subset \mathcal{H}$ que aproxima el espacio de estados relevantes. En el límite de muestreo denso y con la normalización apropiada converge al proyector de Bergman. 

---

## 7. Hamiltoniano efectivo y teoremas espectrales

El Hamiltoniano proyectado es

$$
H_{\text{eff}} = P H P.
$$

Sea $(\mu, v)$ un autopar de $H_{\text{eff}}$, y definamos el residual $r = (H  - \mu)v$.

### Teorema 1 (error espectral)

$$
\mathrm{dist}(\mu, \mathrm{spec}(H)) \le \|(I - P) H P\|.
$$

### Teorema 2 (Davis–Kahan)

Si $\delta$ es la brecha espectral entre subespacios, entonces

$$
\|\sin\Theta(S, U)\| \le \frac{\|(I - P)HP\|}{\delta}.
$$

---

## 8. Evolución temporal y estimación del error

La evolución exacta y proyectada:

$$
\psi(t) = e^{-iHt}\psi_0, \qquad
\tilde\psi(t) = P e^{-iH_{\text{eff}}t}P\psi_0.
$$

### Teorema 3

Si $\|(I - P) H P\| \le \eta$, entonces

$$
\|\psi(t) - \tilde\psi(t)\| \le \eta\|H\| (e^{\|H\|t} - 1).
$$

---

## 9. Geometría topológica y fibrados en el espacio de estados

El espacio de rayos cuánticos $\mathbb{P}(H)$ es isomorfo a $\mathbb{C}P^{d-1}$.  
Cada vector normalizado $\psi \in H$ define un punto $[\psi] \in \mathbb{C}P^{d-1}$.

La **conexión de Berry** en este fibrado (de fase cuántica) es la conexión natural de $\mathcal{O}(-1)$:

$$
A = i \langle \psi, d\psi \rangle, \qquad F = dA = i\, d\langle \psi, d\psi \rangle.
$$

La curvatura $F$ representa la 2-forma de Kähler de Fubini–Study:

$$
F = \omega_{FS}, \qquad \frac{1}{2\pi}\int_{S^2} F \in \mathbb{Z}.
$$

De esta forma, los estados cuánticos se interpretan como **secciones holomorfas del fibrado tautológico**, y sus fases geométricas (fase de Berry) están gobernadas por la topología compleja de $\mathbb{C}P^{d-1}$.

---

## 10. Interpretación en el contexto del modelo de Ising para su implementación programada.

- Los estados coherentes $|Z\rangle$ forman una parametrización holomorfa del espacio de configuraciones magnéticas.  
- El kernel $(1 + z\bar{w})^k$ genera correlaciones y simetrías $SU(2)$ de rotación de spin.  
- La proyección $P$ corresponde a una **reducción twistorial** del sistema de dimensión $2^N$ a un subespacio holomorfo de dimensión $M$, tal que, $M<2^N$.
- De esta manera, $\mathbb{C}P^n$ provee la estructura geométrica natural para espacios de Hilbert cuánticos, mediante una correspondencia topológica diferencial asociando a cada curva un punto proyectivo.  
- El fibrado tautológico y su métrica de Fubini–Study dan lugar a la conexión de Berry y a la estructura Kähler subyacente, pudiendo evaluar el flujo de Ricci.
- La reducción holomorfa, mediante el kernel de Bergman, permite proyectar Hamiltonianos a subespacios coherentes con error controlado.  
- Es útil y también aplicable a sistemas de baja complejidad de entrelazamiento, como los estados coherentes producto y estaod evolucionando a tiempo cortos discretizados en sistemas sin entrelazamiento masivo.
- En este marco, el modelo de Ising puede reinterpretarse como una dinámica sobre secciones de un fibrado holomorfo, lo que abre el camino a descripciones twistoriales y fibradas de sus soluciones cuánticas, también permitiendo la implementación de la cuantización geométrica y de Moyal no conmutativa.

## 1. Solución al sistema.

```python
import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
import matplotlib.pyplot as plt


# Pauli matrices and identity
sx = np.array([[0,1],[1,0]], dtype=complex)
sy=np.array([[0,-1j],[1j,0]], dtype=complex)
sz = np.array([[1,0],[0,-1]], dtype=complex)    
id2 = np.eye(2, dtype=complex)  


# Compute the Kronecker product of a list of matrices
def kron_list_dense(mats):  
    out = mats[0] # Initialize with the first matrix
    for A in mats[1:]:# Iterate over the remaining matrices
        out = np.kron(out, A) # Compute the Kronecker product
    return out 



def kron_list_sparse(mats):
    out = mats[0]
    for A in mats[1:]:
        out = sp.kron(out, A, format='csr')
    return out
```

```python
"""
ising_coherent_projection.py

Pipeline:
 - Construye H sparse de la cadena de Ising 1D.
 - Construye una muestra de estados coherentes por sitio (mapa CP^1: z = tan(theta/2) e^{i phi}).
 - Forma la matriz de datos (cada columna = vector de estado coherente producto).
 - Realiza SVD/Truncación (o SVD del kernel Gram) para obtener proyección P (dim M).
 - Construye H_eff = P^H H P (M x M) y evoluciona psi0 con exp(-i H_eff t).
 - Compara con la evolución completa (si dimension lo permite).
"""

import numpy as np
import scipy.sparse as sp
import scipy.sparse.linalg as spla
from scipy.linalg import svd, eigh, expm
from scipy.sparse.linalg import expm_multiply
import matplotlib.pyplot as plt

# ---------------- Parameters ----------------
N = 8                    # número de espines (prueba N <= 10)
J = 1.0
h = 0.8
periodic = False

# Coherent sampling
S = 300                  # número de estados coherentes muestreados en el fibrado (ajustable)
M_trunc = 40             # dimensión objetivo del subespacio proyectado (M << 2^N)
# Nota: M_trunc <= S y M_trunc <= 2^N. Ajustar según memoria.

# Time evolution
T = 10.0
nt = 201
t_list = np.linspace(0, T, nt)

# ---------------- Pauli (dense & sparse helpers) ----------------
sx = np.array([[0.0,1.0],[1.0,0.0]], dtype=complex)
sy = np.array([[0.0,-1j],[1j,0.0]], dtype=complex)
sz = np.array([[1.0,0.0],[0.0,-1.0]], dtype=complex)
id2 = np.eye(2, dtype=complex)

def kron_list_dense(mats):
    out = mats[0]
    for A in mats[1:]:
        out = np.kron(out, A)
    return out

def kron_list_sparse(mats):
    out = mats[0]
    for A in mats[1:]:
        out = sp.kron(out, A, format='csr')
    return out

# ---------------- Build Ising Hamiltonian (sparse) ----------------
def ising_hamiltonian_sparse(N, J=1.0, h=0.5, periodic=False):
    dim = 2**N
    H = sp.csr_matrix((dim, dim), dtype=complex)
    # interaction zz
    for i in range(N-1 + int(periodic)):
        j = (i+1) % N
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sz)
        ops[j] = sp.csr_matrix(sz)
        H = H - J * kron_list_sparse(ops)
    # transverse field sx
    for i in range(N):
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sx)
        H = H - h * kron_list_sparse(ops)
    return H

# ---------------- Spin-coherent single-site state (holomorphic coordinate z on CP^1) ----------------
def spin_coherent_state_single(z):
    """
    Return single-spin coherent state |z> in computational basis |up>, |down>.
    Using stereographic coordinate z = tan(theta/2) * exp(i phi).
    Normalization: (1 + |z|^2)^{-1/2} [1, z]^T  (holomorphic section representation)
    This is the usual Bargmann-like holomorphic representation for spin-1/2 coherent states.
    """
    norm = 1.0 / np.sqrt(1.0 + np.abs(z)**2)
    vec = np.array([1.0, z], dtype=complex)*norm
    return vec

# ---------------- Product coherent states sample ----------------
def sample_cp1_uniform(S):
    """
    Sample S points approximately uniformly on the sphere S^2 (Haar measure), then map to z.
    Use sampling of cos(theta) uniform and phi uniform.
    """
    u = np.random.uniform(-1.0, 1.0, size=S)   # cos(theta)
    phi = np.random.uniform(0.0, 2*np.pi, size=S)
    theta = np.arccos(u)
    z = np.tan(theta/2.0) * np.exp(1j*phi)     # stereographic coordinate
    return z

def build_product_coherent_vectors(N, z_list):
    """
    Given a list of S stereographic coords z for a single-site coherent state,
    build S many product states in the full 2^N Hilbert space by sampling the SAME z
    for each site (product of identical coherent states) OR randomize per site if desired.
    Two modes:
     - homogeneous product states: same z across all sites (samples sections of the product bundle)
     - heterogeneous (random per site): multiply different z_{s,i} per site (higher coverage)
    We'll give an option: mode='homogeneous' or 'heterogeneous'.
    """
    S = len(z_list)
    dim = 2**N
    Data = np.zeros((dim, S), dtype=complex)

    for k, z in enumerate(z_list):
        # homogeneous product: same single-site coherent state on all sites
        single = spin_coherent_state_single(z)
        # full vector = tensor product of N copies of single
        vec = single
        for _ in range(N-1):
            vec = np.kron(vec, single)
        Data[:, k] = vec
    return Data

# ---------------- Build projection P via SVD (data matrix) ----------------
def build_projection_from_data_matrix(Data, M):
    """
    Given Data (dim x S), compute SVD and take leading M left singular vectors as basis.
    Returns P matrix of shape (dim, M) with orthonormal columns (column basis for subspace).
    Warning: Data can be large; SVD is dense O(dim*S*min(dim,S)).
    """
    # center? not necessary for quantum states; we use raw columns (each normalized)
    # optionally orthonormalize columns first
    U, s, Vh = svd(Data, full_matrices=False)
    P = U[:, :M]   # dim x M
    return P, s

# ---------------- Alternative: SVD of Gram (kernel) ----------------
def build_projection_from_gram(Data, M):
    """
    If dim >> S, it's cheaper to do SVD on the SxS Gram matrix G = Data^H Data.
    Then principal directions in the column space are Data @ eigenvectors(G) / sqrt(eigval).
    """
    S = Data.shape[1]
    G = Data.conj().T @ Data   # S x S
    # Hermitian eigen-decomposition
    evals, evecs = eigh(G)
    # sort descending
    idx = np.argsort(evals)[::-1]
    evals = evals[idx]
    evecs = evecs[:, idx]
    M_eff = min(M, np.sum(evals > 1e-12))
    P = np.zeros((Data.shape[0], M_eff), dtype=complex)
    for m in range(M_eff):
        v = evecs[:, m]
        lam = evals[m]
        # column in original space
        P[:, m] = (Data @ v) / np.sqrt(lam)
    return P, evals

# ---------------- Projected evolution and comparison ----------------
def project_and_evolve(H_sparse, P, psi0, t_list):
    """
    H_sparse: large sparse hermitian (dim x dim)
    P: (dim x M) orthonormal basis columns
    psi0: initial state in full space (dim,)
    Returns: psi_full_t (if feasible via expm_multiply), psi_proj_t (projected evolution lifted back),
             expectation Sz(t) for both.
    We'll compute:
      H_eff = P^H H P  (M x M dense)
      psi0_proj = P^H psi0  (M,)
      evolve psi_eff(t) = exp(-i H_eff t) psi0_proj  (M)  -> lift back with P psi_eff
    """
    # Build H_eff (dense)
    H_eff = (P.conj().T @ (H_sparse @ P)).astype(complex)  # MxM dense
    # Initial projection
    psi0_proj = P.conj().T @ psi0

    # evolve in reduced space: diagonalize H_eff (Hermitian)
    evals_eff, evecs_eff = eigh(H_eff)
    # Precompute time-evolution operator action using diagonalization: exp(-i H_eff t) psi0_proj
    M = H_eff.shape[0]
    psi_eff_t = np.zeros((len(t_list), M), dtype=complex)
    for ti, t in enumerate(t_list):
        phase = np.exp(-1j * evals_eff * t)
        psi_eff_t[ti, :] = evecs_eff @ (phase * (evecs_eff.conj().T @ psi0_proj))

    # lift back to full space
    psi_proj_lifted = (P @ psi_eff_t.T).T   # shape (nt, dim)
    return psi_proj_lifted, H_eff, psi0_proj

# ---------------- Observables helpers ----------------
def build_Sz_total_sparse(N):
    dim = 2**N
    Sz_total = sp.csr_matrix((dim, dim), dtype=complex)
    for i in range(N):
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sz)
        Sz_total += kron_list_sparse(ops)
    return Sz_total

# ---------------- Main flow ----------------
def main():
    # Build Hamiltonian
    H = ising_hamiltonian_sparse(N, J=J, h=h, periodic=periodic)
    dim = 2**N
    print(f"Hamiltoniano construido: N={N}, dim={dim}, nnz={H.nnz}")

    # Sample coherent states on CP^1 (homogeneous product states)
    z_list = sample_cp1_uniform(S)
    Data = build_product_coherent_vectors(N, z_list)
    # Optional: normalize each column (should already be normalized)
    norms = np.linalg.norm(Data, axis=0)
    Data = Data / norms[np.newaxis, :]

    # Build projection P
    # If dim >> S, use Gram method; else direct SVD
    if dim > S:
        print("Usando descomposición de Gram (SxS) para construir P ...")
        P, evals = build_projection_from_gram(Data, M_trunc)
    else:
        print("Usando SVD (dim x S) para construir P ...")
        P, svals = build_projection_from_data_matrix(Data, M_trunc)

    M = P.shape[1]
    print(f"Proyección construida: dim original = {dim}, dimension reducida M = {M}")

    # Initial state psi0 (all up)
    psi0 = np.zeros((dim,), dtype=complex)
    psi0[0] = 1.0

    # Build Sz_total for observables
    Sz_total = build_Sz_total_sparse(N)

    # Project and evolve in reduced space
    psi_proj_lifted, H_eff, psi0_proj = project_and_evolve(H, P, psi0, t_list)

    # Evolve full psi(t) if feasible (exp-multiply)
    print("Computando evolución completa con expm_multiply (si es factible)...")
    try:
        A = -1j * H
        psi_full_t = expm_multiply(A, psi0, start=0.0, stop=T, num=len(t_list))
    except Exception as e:
        print("Evolución completa no factible por memoria/tiempo:", e)
        psi_full_t = None

    # Compute expectation values <Sz>/sitio for ambas evoluciones
    Sz_mat = Sz_total  # sparse
    exp_proj = np.empty(len(t_list), dtype=float)
    exp_full = None if psi_full_t is None else np.empty(len(t_list), dtype=float)
    for k in range(len(t_list)):
        v = psi_proj_lifted[k]
        exp_proj[k] = (v.conj() @ (Sz_mat @ v)).real / N
        if psi_full_t is not None:
            u = psi_full_t[k]
            exp_full[k] = (u.conj() @ (Sz_mat @ u)).real / N

    # Plot comparison
    plt.figure(figsize=(8,4))
    plt.plot(t_list, exp_proj, label='proyectado (lifted)')
    if psi_full_t is not None:
        plt.plot(t_list, exp_full, label='completo', linestyle='--')
    plt.xlabel('t')
    plt.ylabel(r'$\langle S^z \rangle$ por sitio')
    plt.legend()
    plt.title(f'Evolución proyectada vs completa (N={N}, M={M})')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Print diagnostics: fidelity inicial entre psi0 y su proyección
    psi0_reconstructed = P @ psi0_proj
    fid0 = np.abs(np.vdot(psi0_reconstructed, psi0))**2
    print(f"Fidelidad inicial entre psi0 y P P^H psi0: {fid0:.6e}")
    # Error medio de magnetización si full disponible
    if psi_full_t is not None:
        err = np.linalg.norm(exp_full - exp_proj) / np.linalg.norm(exp_full)
        print(f"Error relativo L2 en magnetización: {err:.3e}")

    # Save reduced H_eff and P if needed
    # np.savez('reduction_P_Heff.npz', P=P, H_eff=H_eff, psi0_proj=psi0_proj, evals=evals if 'evals' in locals() else None)
    return

if __name__ == "__main__":
    main()
```

# 2. Sobre el Flujo de Ricci 
La fidelidad $\mathcal{F}$ mide qué tan bien el estado inicial $\psi_0$ está contenido (o capturado) dentro del subespacio de baja dimensión $M$ que se construye en cada paso del flujo, pudiendo implementarse de la siguiente manera,

La función utilizada calcula 
$$
\mathcal{F} = \left| \langle \psi_0 | P P^H | \psi_0 \rangle \right|^2
$$

Donde
- $\psi_0 = |00...0\rangle$ es el estado inicial (todo \textit{up} en el código).
- $P$ es la matriz de proyección (de dimensión $D \times M$), cuyas columnas forman la base ortonormal del subespacio de dimensión $M$.
- $P P^H$ es el operador proyector sobre el subespacio $M$.

Si $\mathcal{F} \approx 1$, el subespacio $M$ es excelente para representar $\psi_0$. En cambio, si $\mathcal{F} \approx 0$, el subespacio $M$ es una mala elección de base para $\psi_0$.

---

## 2.1. Parámetro $\beta$ como escala geométrica.

El Flujo de Ricci se simula al variar $\beta$, que actúa como una \textbf{escala de resolución geométrica} o un tiempo de flujo ($\tau$) en el contexto de la teoría de campos.

El proyector $P(\beta)$ se construye utilizando el \textbf{Kernel Ponderado}, donde los pesos $w_k$ son proporcionales a

$$
w_k(\beta) \propto \langle Z_k | e^{-\beta H} | Z_k \rangle
$$

- \textbf{$\beta \approx 0$ (Alta Temperatura / Escala Larga):} Cuando $\beta$ es muy pequeño (alta temperatura), $e^{-\beta H} \approx \mathbb{I}$ (la identidad). Los pesos $w_k$ son casi uniformes. El subespacio $P(0)$ se define por la métrica geométrica "desnuda" sin tener en cuenta la dinámica del Hamiltoniano $H$.
- \textbf{$\beta$ grande (Baja Temperatura / Escala Corta):} $e^{-\beta H}$ proyecta fuertemente sobre los autoestados de baja energía de $H$. El proyector $P(\beta)$ da más peso a los estados coherentes $|Z_k\rangle$ que están cerca del \textit{manifold} de baja energía de $H$. Es la métrica geométrica "vestida", adaptada a la dinámica del sistema.

---

## 2.2. Interpretación de la curva $\mathcal{F}(\beta)$


La gráfica de $\mathcal{F}$ vs. $\beta$ constituye una herramienta diagnóstica fundamental, pues revela la eficiencia y la convergencia del flujo de Ricci hacia el subespacio de baja energía invariante definido por el Hamiltoniano $H$. La función de fidelidad $\mathcal{F}: \beta \mapsto [0, 1]$ se analiza rigurosamente mediante la división en tres regímenes distintos, análogos a las escalas en la física térmica o el flujo del grupo de renormalización.

### 2.2.1. $\beta \approx 0$: Límite Geométrico y de Altas Energías

En el límite de $\beta \to 0$, la temperatura efectiva $T \sim 1/\beta$ es infinita, siendo un límite geométrico y de altas energías. El factor de ponderación del kernel, $w_k(\beta) \propto \langle Z_k | e^{-\beta H} | Z_k \rangle$, se aproxima a la identidad, $w_k(\beta) \approx \langle Z_k | \mathbb{I} | Z_k \rangle = 1$, asumiendo la normalización del \textit{manifold}. Consecuentemente, el proyector $P(0)$ se construye exclusivamente a partir de la estructura geométrica del \textit{manifold} de estados coherentes, desacoplado de las propiedades espectrales de $H$. La fidelidad $\mathcal{F}(0)$ es típicamente baja o moderada, ya que el subespacio $M$ resultante es una representación puramente geométrica, no optimizada para capturar las componentes de baja energía del estado inicial $\psi_0$.

### 2.2.2. $\beta$ Intermedio: Régimen de Flujo y Optimización Métrica


A medida que $\beta$ crece desde cero, el operador exponencial $e^{-\beta H}$ comienza a diferenciar las ponderaciones $w_k(\beta)$, siendo un régimen de energías de flujo y optimización métrica. Los estados $|Z_k\rangle$ que poseen una mayor superposición con los autoestados de menor energía de $H$ (aquellos con autovalores $\lambda_n$ pequeños, haciendo $e^{-\beta \lambda_n}$ grande) experimentan una amplificación exponencial en sus pesos. Físicamente, esto indica que la estructura métrica del subespacio cuántico está siendo activamente \textit{vestida} o \textit{adaptada} por la dinámica hamiltoniana. El rápido aumento de $\mathcal{F}(\beta)$ en esta región demuestra la eficiencia del ponderado de tipo Flujo de Ricci para rotar a la sub-variedad diferencial $M$ hacia el subespacio de baja energía dinámicamente relevante, logrando una captura significativamente mejor del estado inicial $\psi_0$.

### 2.2.3. $\beta$ Grande: Límite del Estado Fundamental y de Baja Energía


En el límite de $\beta \to \infty$ (temperatura cero), $e^{-\beta H}$ se comporta asintóticamente como un proyector perfecto sobre el estado fundamental $\psi_g$ o el subespacio degenerado de baja energía. El kernel ponderado filtra rigurosamente las contribuciones de alta energía, lo que obliga a las columnas de $P(\beta)$ a abarcar el subespacio de menor energía de $H$. La estabilización de la fidelidad, $\mathcal{F}(\beta) \to \mathcal{F}_{\text{max}} \le 1$, señala la convergencia exitosa del sub-manifold $M$ al subespacio invariante verdadero de $H$. Si el estado inicial $\psi_0$ tiene una fuerte superposición con este \textit{manifold} de baja energía, $\mathcal{F}_{\text{max}}$ tenderá a la unidad, confirmando que el sub-manifold $M$ proporciona una representación óptima y dinámicamente adaptada de la física a largo plazo del sistema.

En resumen, la curva $\mathcal{F}(\beta)$ describe, ¿qué tan rápido y qué tan bien la geometría del subespacio, guiada por el Hamiltoniano $H$ (a través del Kernel Ponderado), se auto-adapta para capturar el estado $\psi_0$? 

Un aumento pronunciado significa que el flujo de $\beta$ refina eficientemente el subespacio hacia la región de interés.

```python
import numpy as np
import scipy.sparse as sp
from scipy.linalg import eigh, expm
import matplotlib.pyplot as plt
import time

# ---------------- Parámetros del Sistema y Muestreo ----------------
N = 8               # Number of spins
J = 1.0             
h = 0.8             
S = 500             # Number of sampled coherent states
M_trunc = 50        # Target dimension
BETA_MAX = 5.0      # Maximum geometric scale (Ricci Flow end point)

# ---------------- Operadores y Funciones Auxiliares MÍNIMAS (Copias esenciales) ----------------

sx = np.array([[0.0,1.0],[1.0,0.0]], dtype=complex)
sz = np.array([[1.0,0.0],[0.0,-1.0]], dtype=complex)

def kron_list_sparse(mats):
    """Calculates the Kronecker product of a list of sparse matrices."""
    out = mats[0]
    for A in mats[1:]:
        out = sp.kron(out, A, format='csr')
    return out

def ising_hamiltonian_sparse(N, J=1.0, h=0.5, periodic=False):
    """Builds the Ising Hamiltonian (sparse)."""
    dim = 2**N
    H = sp.csr_matrix((dim, dim), dtype=complex)
    # ZZ interaction
    for i in range(N-1 + int(periodic)):
        j = (i+1) % N
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sz); ops[j] = sp.csr_matrix(sz)
        H = H - J * kron_list_sparse(ops)
    # Transverse field X
    for i in range(N):
        ops = [sp.eye(2, format='csr', dtype=complex) for _ in range(N)]
        ops[i] = sp.csr_matrix(sx)
        H = H - h * kron_list_sparse(ops)
    return H

def spin_coherent_state_single(z):
    """Returns single-site coherent state |z>."""
    norm = 1.0 / np.sqrt(1.0 + np.abs(z)**2)
    return np.array([1.0, z], dtype=complex) * norm

def sample_cp1_uniform(count):
    """Samples 'count' stereographic coordinates (z) uniformly on S^2 (Haar measure)."""
    u = np.random.uniform(-1.0, 1.0, size=count)
    phi = np.random.uniform(0.0, 2*np.pi, size=count)
    theta = np.arccos(u)
    z = np.tan(theta/2.0) * np.exp(1j*phi)
    return z

def build_product_coherent_vectors_optimized(N, S):
    """
    Builds the Data matrix D using heterogeneous (optimized) product states: 
    |Z_k> = |z_k1> \otimes ... \otimes |z_kN> (different z for each site).
    """
    dim = 2**N
    Data = np.zeros((dim, S), dtype=complex)
    z_samples_all = sample_cp1_uniform(S * N) 
    for k in range(S):
        z_sample_k = z_samples_all[k*N : (k+1)*N] 
        vec = spin_coherent_state_single(z_sample_k[0])
        for i in range(1, N):
            vec = np.kron(vec, spin_coherent_state_single(z_sample_k[i]))
        Data[:, k] = vec
    return Data

# ---------------- Kernel Ponderado (Esencial para el Flujo) ----------------

def get_thermal_weights(H_sparse, Data, beta):
    """
    Calculates weights w_k = <Z_k | exp(-beta H) | Z_k>, modeling the Hamiltonian structure.
    """
    if beta < 1e-6:
        return np.ones(Data.shape[1], dtype=float)
        
    H_dense = H_sparse.todense()
    P_op = expm(-beta * H_dense)
    weights = np.zeros(Data.shape[1], dtype=float)
    
    for k in range(Data.shape[1]):
        Zk = Data[:, k]
        weights[k] = np.real(Zk.conj() @ (P_op @ Zk))
        
    weights /= np.sum(weights)
    return weights

def build_projection_from_weighted_data(Data, M, weights):
    """
    Builds the Projection matrix P using the weighted Gram kernel (D_tilde^H D_tilde).
    """
    sqrt_weights = np.sqrt(weights)
    D_tilde = Data * sqrt_weights[np.newaxis, :]
    
    G_tilde = D_tilde.conj().T @ D_tilde 
    
    evals, evecs = eigh(G_tilde)
    idx = np.argsort(evals)[::-1]
    evals = evals[idx]; evecs = evecs[:, idx]
    M_eff = min(M, np.sum(evals > 1e-12))
    
    P = np.zeros((Data.shape[0], M_eff), dtype=complex)
    for m in range(M_eff):
        v = evecs[:, m]
        lam = evals[m]
        P[:, m] = (D_tilde @ v) / np.sqrt(lam)
        
    return P, evals

# ---------------- Flujo de Ricci (Core Logic) ----------------

def calculate_initial_fidelity(P_matrix, psi0):
    """
    Calculates the initial fidelity F = |<psi0| P P^H |psi0>|^2.
    """
    psi0_proj = P_matrix.conj().T @ psi0
    psi0_reconstructed = P_matrix @ psi0_proj
    fidelity = np.abs(np.vdot(psi0_reconstructed, psi0))**2
    return fidelity

def analyze_ricci_flow_path(H, Data, M_trunc, beta_list):
    """
    Analyzes the 'Ricci Flow' path by varying the geometric scale parameter beta.
    """
    dim = H.shape[0]
    psi0 = np.zeros((dim,), dtype=complex)
    psi0[0] = 1.0  # Initial state |00...0>
    
    fidelities = []
    
    print("\n--- Análisis de Flujo de Ricci (Escala Beta) ---")
    
    for beta in beta_list:
        start_time = time.time()
        
        # 1. Get the thermal weights (w_k)
        weights = get_thermal_weights(H, Data, beta)
        
        # 2. Build the weighted Projection P(beta)
        P_beta, _ = build_projection_from_weighted_data(Data, M_trunc, weights)
        
        # 3. Calculate the Fidelity (Quality metric)
        fid = calculate_initial_fidelity(P_beta, psi0)
        fidelities.append(fid)
        
        print(f"Beta (Escala $\\tau$)={beta:.2f}: Fidelidad={fid:.6f} | Tiempo: {time.time() - start_time:.2f}s")
        
    return np.array(fidelities)

# ---------------- Main Flow ----------------
def main_ricci_flow_standalone():
    """Entry point for the standalone Ricci Flow analysis."""
    # Build Hamiltonian
    H = ising_hamiltonian_sparse(N, J=J, h=h)
    dim = 2**N
    print(f"Sistema: N={N}, Dim={dim}, M={M_trunc}, Muestras={S}")
    
    # 1. Sample heterogeneous data
    Data = build_product_coherent_vectors_optimized(N, S)
    
    # 2. Beta scale list (simulating the flow trajectory)
    beta_list = np.linspace(0.01, BETA_MAX, 10) 
    
    # 3. Run the 'Flow' analysis
    fidelities = analyze_ricci_flow_path(H, Data, M_trunc, beta_list)
    
    # 4. Visualize the quality parameter evolution
    plt.figure(figsize=(9, 5), dpi=100)
    plt.plot(beta_list, fidelities, marker='o', color='#8b5cf6', linestyle='-', linewidth=2)
    plt.title(r'Evolución de la Fidelidad $\mathcal{F}(\beta)$ (Análogo al Flujo de Ricci)', fontsize=14)
    plt.xlabel(r'Escala Geométrica $\beta$ (Tiempo de Flujo $\tau$)', fontsize=12)
    plt.ylabel(r'Fidelidad Inicial $\mathcal{F}$', fontsize=12)
    plt.grid(True, alpha=0.5, linestyle=':')
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main_ricci_flow_standalone()
```
